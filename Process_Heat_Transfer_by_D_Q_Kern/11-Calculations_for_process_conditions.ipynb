# Chapter 11: Calculations for process conditions

## Example 11.1: Calculation_of_a_Straw_Oil_Naphtha_Exchanger.sce

```scilab
printf('	 example 11.1 
');
printf('	 approximate values are mentioned in the book 
');
T1=340; // inlet hot fluid,F
T2=240; // outlet hot fluid,F
t1=200; // inlet cold fluid,F
t2=230; // outlet cold fluid,F
W=29800; // lb/hr
w=103000; // lb/hr
printf('	 1.for heat balance 
');
printf('	 for straw oil 
');
c=0.58; // Btu/(lb)*(F)
Q=((W)*(c)*(T1-T2)); // Btu/hr
printf('	 total heat required for straw oil is : %.2e Btu/hr 
',Q);
printf('	 for naphtha 
');
c=0.56; // Btu/(lb)*(F)
Q=((w)*(c)*(t2-t1)); // Btu/hr
printf('	 total heat required for naphtha is : %.2e Btu/hr 
',Q);
delt1=T2-t1; //F
delt2=T1-t2; // F
printf('	 delt1 is : %.0f F 
',delt1);
printf('	 delt2 is : %.0f F 
',delt2);
LMTD=((delt2-delt1)/((2.3)*(log10(delt2/delt1))));
printf('	 LMTD is :%.1f F 
',LMTD);
R=((T1-T2)/(t2-t1));
printf('	 R is : %.1f 
',R);
S=((t2-t1)/(T1-t1));
printf('	 S is : %.3f 
',S);
printf('	 FT is 0.885 
'); // from fig 18
delt=(0.885*LMTD); // F
printf('	 delt is : %.1f F 
',delt);
X=((delt1)/(delt2));
printf('	 ratio of two local temperature difference is : %.3f 
',X);
L=16;
Fc=0.405; // from fig.17
Kc=0.23; // crude oil controlling
Tc=((T2)+((Fc)*(T1-T2))); // caloric temperature of hot fluid,F
printf('	 caloric temperature of hot fluid is : %.1f F 
',Tc);
tc=((t1)+((Fc)*(t2-t1))); // caloric temperature of cold fluid,F
printf('	 caloric temperature of cold fluid is : %.0f F 
',tc);
UD1=70; // assume, from table 8a
A1=((Q)/((UD1)*(delt)));
printf('	 A1 is : %.0f ft^2 
',A1);
a1=0.1963; // ft^2/lin ft
N1=(A1/(16*a1));
printf('	 number of tubes are : %.0f 
',N1);
N2=124; // assuming two tube passes, from table 9
A2=(N2*L*a1); // ft^2
printf('	 total surface area is : %.1e ft^2 
',A2);
UD=((Q)/((A2)*(delt)));
printf('	 correct design overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UD);
printf('	 hot fluid:shell side,straw oil 
');
ID=15.25; // in
C=0.25; // clearance
B=3.5; // minimum baffle spacing,from eq 11.4,in
PT=1;
as=((ID*C*B)/(144*PT)); // flow area,from eq 7.1,ft^2
printf('	 flow area is : %.4f ft^2 
',as);
Gs=(W/as); // mass velocity,from eq 7.2,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gs);
mu1=3.63; // at 280.5F,lb/(ft)*(hr), from fig.14
De=0.95/12; // from fig.28,ft
Res=((De)*(Gs)/mu1); // reynolds number
printf('	 reynolds number is : %.0e 
',Res);
jH=46; // from fig.28
Z=0.224; // Z=(K*(c*mu3/k)^(1/3)),Btu/(hr)(ft^2)(F/ft), at mu3=1.5cp and 35 API
Ho=((jH)*(1/De)*(Z)); // H0=(h0/phya),using eq.6.15,Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Ho);
phys=1;
ho=(Ho)*(phys); // from eq.6.36
printf('	 Correct h0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',ho);
printf('	 cold fluid:inner tube side,naphtha 
');
Nt=124;
n=2; // number of passes
L=16; //ft
at1=0.302; // flow area, in^2
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=(w/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gt);
mu2=1.31; // at 212F,lb/(ft)*(hr)
D=0.0517; // ft
Ret=((D)*(Gt)/mu2); // reynolds number
printf('	 reynolds number is : %.2e 
',Ret);
jH=102; // from fig.24
Z=0.167; // Z=(K*(c*mu3/k)^(1/3)),Btu/(hr)(ft^2)(F/ft), at mu4=0.54cp and 48 API
Hi=((jH)*(1/D)*(Z)); //Hi=(hi/phyp),using eq.6.15a,Btu/(hr)*(ft^2)*(F)
printf('	 Hi is : %.0f Btu/(hr)*(ft^2)*(F) 
',Hi);
ID=0.62; // ft
OD=0.75; //ft
Hio=((Hi)*(ID/OD)); //Hio=(hio/phyp), using eq.6.5
printf('	 Correct Hi0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',Hio);
phyt=1;
hio=(Hio)*(phyt); // from eq.6.37
printf('	 Correct hi0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',hio);
printf('	 pressure drop  for annulus 
');
f=0.00225; // friction factor for reynolds number 7000, using fig.29
s=0.76; // for reynolds number 7000,using fig.6
Ds=15.25/12; // ft
N=(12*L/B); // number of crosses,using eq.7.43
printf('	 number of crosses are : %.0f 
',N);
delPs=((f*(Gs^2)*(Ds)*(N))/(5.22*(10^10)*(De)*(s)*(phys))); // using eq.7.44,psi
printf('	 delPs is : %.1f psi 
',delPs);
printf('	 pressure drop  for inner pipe 
');
f=0.0002; // friction factor for reynolds number 31300, using fig.26
s=0.72;
delPt=((f*(Gt^2)*(L)*(n))/(5.22*(10^10)*(D)*(s)*(phyt))); // using eq.7.45,psi
printf('	 delPt is : %.1f psi 
',delPt);
Uc=((hio)*(ho)/(hio+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',Uc);
Rd=((Uc-UD)/((UD)*(Uc))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.4f (hr)*(ft^2)*(F)/Btu 
',Rd);
printf('	 The first trial is disqualified because of failure to meet the required dirt factor 
');
printf('	 Proceeding as above and carrying the viscosity correction and pressure drops to completion the new summary is given using a 17.25in. ID shell with 166 tubes on two passes and a 3.5in. baffle space 
');
UD1=60; // assumption for 2 tube passes,3.5 baffle spacing and 17.25in ID
UC1=74.8;
printf('	 clean overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UC1);
UD2=54.2;
printf('	 correct design overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UD2);
Rd1=0.005; 
printf('	 actual Rd is : %.3f (hr)*(ft^2)*(F)/Btu 
',Rd1);
delPs1=4.7;
printf('	 delPs is : %.1f psi 
',delPs1);
delPt1=2.1;
printf('	 delPt is : %.1f psi 
',delPt1);
//end
```

## Example 11.2: Calculation_of_a_Flue_Gas_Cooler.sce

```scilab
printf('	 example 11.2 
');
printf('	 approximate values are mentioned in the book 
');   
T1=350; // inlet hot fluid,F
T2=160; // outlet hot fluid,F
t1=100; // inlet cold fluid,F
t2=295; // outlet cold fluid,F
W=84438; // lb/hr
w=86357; // lb/hr
printf('	 1.for heat balance 
');
printf('	 for lean oil 
');
c=0.56; // Btu/(lb)*(F)
Qh=((W)*(c)*(T1-T2)); // Btu/hr
printf('	 total heat required for lean oil is : %.2e Btu/hr 
',Qh);
printf('	 for rich oil 
');
c=0.53; // Btu/(lb)*(F)
Qc=((w)*(c)*(t2-t1)); // Btu/hr
printf('	 total heat required for rich oil is : %.2e Btu/hr 
',Qc);
Q=(Qh+Qc)/(2);
printf('	 Q is : %.2e V 
',Q);
delt1=T2-t1; //F
delt2=T1-t2; // F
printf('	 delt1 is : %.0f F 
',delt1);
printf('	 delt2 is : %.0f F 
',delt2);
LMTD=((delt2-delt1)/((2.3)*(log10(delt2/delt1))));
printf('	 LMTD is :%.1f F 
',LMTD);
R=((T1-T2)/(t2-t1));
printf('	 R is : %.3f 
',R);
S=((t2-t1)/(T1-t1));
printf('	 S is : %.2f 
',S);
printf('	 FT is 0.875 
'); // for 4-8 exchanger,from fig 21
delt=(0.875*LMTD); // F
printf('	 delt is : %.1f F 
',delt);
X=((delt1)/(delt2));
printf('	 ratio of two local temperature difference is : %.2f 
',X);
Fc=0.48; // from fig.17
Kc=0.32; // crude oil controlling
Tc=((T2)+((Fc)*(T1-T2))); // caloric temperature of hot fluid,F
printf('	 caloric temperature of hot fluid is : %.0f F 
',Tc);
tc=((t1)+((Fc)*(t2-t1))); // caloric temperature of cold fluid,F
printf('	 caloric temperature of cold fluid is : %.1f 
',tc);
UD1=50; // assume, from table 8a
A1=((Q)/((UD1)*(delt)));
printf('	 A1 is : %.2e ft^2 
',A1);
a1=0.1963; // ft^2/lin ft
N1=(A1/(16*a1*2)); // 2-4 exchanger in series
printf('	 number of tubes are : %.0f 
',N1);
N2=580; // assuming six tube passes,31in ID, from table 9
A2=(N2*16*a1*2); // ft^2
printf('	 total surface area is : %.2e ft^2 
',A2);
UD=((Q)/((A2)*(delt)));
printf('	 correct design overall coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',UD);
printf('	 hot fluid:inner tube side,lean oil 
');
Nt=580;
n=6; // number of passes
L=16; //ft
at1=0.302; // flow area, in^2
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=(W/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gt);
mu2=2.13; // at 212F,lb/(ft)*(hr)
D=0.0517; // ft
Ret=((D)*(Gt)/mu2); // reynolds number
printf('	 reynolds number is : %.2e 
',Ret);
jH=36.5; // from fig.24
Z=0.185; // Z=(K*(c*mu3/k)^(1/3)),Btu/(hr)(ft^2)(F/ft), at mu4=0.88cp and 35 API
Hi=((jH)*(1/D)*(Z)); //Hi=(hi/phyp),using eq.6.15a,Btu/(hr)*(ft^2)*(F)
printf('	 Hi is : %.0f Btu/(hr)*(ft^2)*(F) 
',Hi);
ID=0.62; // ft
OD=0.75; //ft
Hio=((Hi)*(ID/OD)); //Hio=(hio/phyp), using eq.6.5
printf('	 Correct Hi0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',Hio);
phyt=1;
hio=(Hio)*(phyt); // from eq.6.37
printf('	 Correct hi0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',hio);
printf('	 cold fluid:shell side,rich oil 
');
ID=31; // in
C=0.25; // clearance
B=12; // minimum baffle spacing,from eq 11.4,in
PT=1;
as=((ID*C*B)/(144*PT))/(2); // flow area,from eq 7.1,ft^2
printf('	 flow area is : %.3f ft^2 
',as);
Gs=(w/as); // mass velocity,from eq 7.2,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gs);
mu1=3.15; // at 193.5F,lb/(ft)*(hr), from fig.14
De=0.95/12; // from fig.28,ft
Res=((De)*(Gs)/mu1); // reynolds number
printf('	 reynolds number is : %.2e 
',Res);
jH=45; // from fig.28
Z=0.213; // Z=(K*(c*mu3/k)^(1/3)),Btu/(hr)(ft^2)(F/ft), at mu3=1.3cp and 35 API
Ho=((jH)*(1/De)*(Z)); // H0=(h0/phya),using eq.6.15,Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Ho);
phys=1;
ho=(Ho)*(phys); // from eq.6.36
printf('	 Correct h0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',ho);
printf('	 pressure drop  for inner pipe 
');
f=0.00027; // friction factor for reynolds number 10100, using fig.26
s=0.77;
delPt=((2*f*(Gt^2)*(L)*(n))/(5.22*(10^10)*(D)*(s)*(phyt))); // using eq.7.45,psi
printf('	 delPt is : %.1f psi 
',delPt);
X1=0.024; // X1=((V^2)/(2*g)), for Gt 1060000,using fig.27
delPr=((4*2*n*X1)/(s)); // using eq.7.46,psi
printf('	 delPr is : %.1f psi 
',delPr);
delPT=delPt+delPr; // using eq.7.47,psi
printf('	 delPT is : %.1f psi 
',delPT);
printf('	 allowable delPT is 10 psi 
');
printf('	 pressure drop  for annulus 
');
f=0.0023; // friction factor for reynolds number 6720, using fig.29
s=0.79; // for reynolds number 6720,using fig.6
Ds=31/12; // ft
De=0.0792;
N=(4*12*L/B); // number of crosses,using eq.7.43
printf('	 number of crosses are : %.0f 
',N);
delPs=((f*(Gs^2)*(Ds)*(N))/(5.22*(10^10)*(De)*(s)*(phys))); // using eq.7.44,psi
printf('	 delPs is : %.1f psi 
',delPs);
printf('	 allowable delPa is 10 psi 
');
Uc=((hio)*(ho)/(hio+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',Uc);
Rd=((Uc-UD)/((UD)*(Uc))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.4f (hr)*(ft^2)*(F)/Btu 
',Rd);
printf('	 The initial assumptions have provided an exchanger which very nearly meets all the requirements. Eight-pass units would meet the heat-transfer requirement but would give a tube-side pressure drop of 14 psi. The trial exchanger will be somewhat less suitable when the value of Q, is also taken into account. If the minimum dirt factor of 0.0040 is to be taken literally, it will be necessary to try the next size shell 
');
printf('	 Assume a 33 in. ID shell with six1 tube passes and baffies spaced 12-in. apart, since the pressure drop increases with the diameter of the shell for a given mass velocity. 
');
UC1=52.3;
printf('	 clean overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UC1);
UD2=42;
printf('	 correct design overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UD2);
Rd1=0.0047; 
printf('	 calculated Rd is : %.4f (hr)*(ft^2)*(F)/Btu 
',Rd1);
Rd2=0.004; 
printf('	 actual Rd is : %.3f (hr)*(ft^2)*(F)/Btu 
',Rd2);
delPs1=4.4;
printf('	 delPs is : %.1f psi 
',delPs1);
delPT1=7.9;
printf('	 delPt is : %.1f psi 
',delPT1);
//end
```

## Example 11.3: Calculation_of_a_Caustic_Solution_Cooler.sce

```scilab
printf('	 example 11.3 
');
printf('	 approximate values are mentioned in the book 
');
T1=190; // inlet hot fluid,F
T2=120; // outlet hot fluid,F
t1=80; // inlet cold fluid,F
t2=120; // outlet cold fluid,F
W=100000; // lb/hr
w=154000; // lb/hr
printf('	 1.for heat balance 
');
printf('	 for caustic 
');
c=0.88; // Btu/(lb)*(F)
Q=((W)*(c)*(T1-T2)); // Btu/hr
printf('	 total heat required for caustic is : %.2e Btu/hr 
',Q);
printf('	 for water 
');
c=1; // Btu/(lb)*(F)
Q=((w)*(c)*(t2-t1)); // Btu/hr
printf('	 total heat required for water is : %.2e Btu/hr 
',Q);
delt1=T2-t1; //F
delt2=T1-t2; // F
printf('	 delt1 is : %.0f F 
',delt1);
printf('	 delt2 is : %.0f F 
',delt2);
LMTD=((delt2-delt1)/((2.3)*(log10(delt2/delt1))));
printf('	 LMTD is :%.1f F 
',LMTD);
R=((T1-T2)/(t2-t1));
printf('	 R is : %.2f 
',R);
S=((t2-t1)/(T1-t1));
printf('	 S is : %.3f 
',S);
printf('	 FT is 0.815 
'); // for 4-8 exchanger,from fig 21
delt=(0.815*LMTD); // F
printf('	 delt is : %.1f F 
',delt);
Tc=((T2)+(T1))/(2); // caloric temperature of hot fluid,F
printf('	 caloric temperature of hot fluid is : %.0f F 
',Tc);
tc=((t1)+(t2))/(2); // caloric temperature of cold fluid,F
printf('	 caloric temperature of cold fluid is : %.1f 
',tc);
UD1=250; // assume, from table 8
A1=((Q)/((UD1)*(delt)));
printf('	 A1 is : %.0f ft^2 
',A1);
a1=0.2618; // ft^2/lin ft
L=16;
N1=(A1/(16*a1));
printf('	 number of tubes are : %.0f 
',N1);
N2=140; // assuming four tube passes,19.25in ID, from table 9
A2=(N2*L*a1); // ft^2
printf('	 total surface area is : %.0f ft^2 
',A2);
UD=((Q)/((A2)*(delt)));
printf('	 correct design overall coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',UD);
printf('	 hot fluid:shell side,caustic 
');
ID=19.25; // in
C=0.25; // clearance
B=7; // minimum baffle spacing,from eq 11.4,in
PT=1.25;
as=((ID*C*B)/(144*PT)); // flow area,from eq 7.1,ft^2
printf('	 flow area is : %.4f ft^2 
',as);
Gs=(W/as); // mass velocity,from eq 7.2,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gs);
mu1=1.84; // at 155F,lb/(ft)*(hr), from fig.14
De=0.72/12; // from fig.28,ft
Res=((De)*(Gs)/mu1); // reynolds number
printf('	 reynolds number is : %.2e 
',Res);
jH=75; // from fig.28
Z=0.575; // Z=(K*(c*mu3/k)^(1/3)),Btu/(hr)(ft^2)(F/ft)
Ho=((jH)*(1/De)*(Z)); // H0=(h0/phya),using eq.6.15,Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Ho);
phys=1; // low viscosity
ho=(Ho)*(phys); // from eq.6.36
printf('	 Correct h0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',ho);
printf('	 cold fluid:inner tube side,water 
');
Nt=140;
n=4; // number of passes
L=16; //ft
at1=0.546; // flow area, in^2
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=(w/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gt);
V=Gt/(3600*62.5);
printf('	 V is %.2f fps 
',V);
mu2=1.74; // at 100F,lb/(ft)*(hr)
D=0.0695; // ft
Ret=((D)*(Gt)/mu2); // reynolds number
printf('	 reynolds number is : %.2e 
',Ret);
hi=1240*0.94; // from fig 25
printf('	 Hi is : %.0f Btu/(hr)*(ft^2)*(F) 
',hi);
ID=0.834; // ft
OD=1; //ft
hio=((hi)*(ID/OD)); //Hio=(hio/phyp), using eq.6.5
printf('	 Correct hi0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',hio);
printf('	 pressure drop  for annulus 
');
f=0.0019; // friction factor for reynolds number 17400, using fig.29
s=1.115; // for reynolds number 17400,using fig.6
Ds=19.25/12; // ft
De=0.06;
N=(12*L/B)+1; // number of crosses,using eq.7.43
printf('	 number of crosses are : %.0f 
',N);
delPs=((f*(Gs^2)*(Ds)*(N))/(5.22*(10^10)*(De)*(s)*(phys))); // using eq.7.44,psi
printf('	 delPs is : %.0f psi 
',delPs);
printf('	 allowable delPa is 10 psi 
');
printf('	 pressure drop  for inner pipe 
');
f=0.00018; // friction factor for reynolds number 46300, using fig.26
s=1;
phyt=1;
delPt=((f*(Gt^2)*(L)*(n))/(5.22*(10^10)*(D)*(s)*(phyt))); // using eq.7.45,psi
printf('	 delPt is : %.1f psi 
',delPt);
X1=0.18; // X1=((V^2)/(2*g)), for Gt 1060000,using fig.27
delPr=((4*n*X1)/(s)); // using eq.7.46,psi
printf('	 delPr is : %.1f psi 
',delPr);
delPT=delPt+delPr; // using eq.7.47,psi
printf('	 delPT is : %.1f psi 
',delPT);
printf('	 allowable delPa is 10 psi 
');
Uc=((hio)*(ho)/(hio+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Uc);
Rd=((Uc-UD)/((UD)*(Uc))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.4f (hr)*(ft^2)*(F)/Btu 
',Rd);
printf('	 Adjustment of the baffie space to use the full 10 psi will still not permit the exchanger to make the 0.002 dirt factor. The value of UD has been assumed too high 
');
printf('	 Try a 21.25in.ID shell with four tube passes and a 6in.baffie·space.This corresponds to 170 tubes 
');
UC1=390;
printf('	 clean overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UC1);
UD2=200;
printf('	 correct design overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UD2);
Rd1=0.0024; 
printf('	 calculated Rd is : %.4f (hr)*(ft^2)*(F)/Btu 
',Rd1);
Rd2=0.002; 
printf('	 actual Rd is : %.3f (hr)*(ft^2)*(F)/Btu 
',Rd2);
delPs1=9.8;
printf('	 delPs is : %.1f psi 
',delPs1);
delPT1=4.9;
printf('	 delPt is : %.1f psi 
',delPT1);
//end
```
