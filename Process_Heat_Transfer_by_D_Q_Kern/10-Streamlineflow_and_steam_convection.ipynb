## Example 10.1: Crude_Oil_Heater.sce

```scilab
printf('	 example 10.1 
');
printf('	 approximate values are mentioned in the book 
');
T1=250; // inlet hot fluid,F
T2=250; // outlet hot fluid,F
t1=95; // inlet cold fluid,F
t2=145; // outlet cold fluid,F
W=16000; // lb/hr
w=410; // lb/hr
printf('	 1.for heat balance 
');
printf('	 for crude 
');
c=0.485; // Btu/(lb)*(F)
Q=((W)*(c)*(t2-t1)); // Btu/hr
printf('	 total heat required for crude is : %.2e Btu/hr 
',Q);
printf('	 for steam 
');
l=945.5; // Btu/(lb)
Q=((w)*(l)); // Btu/hr
printf('	 total heat required for steam is : %.2e Btu/hr 
',Q);
delt1=T2-t1; //F
delt2=T1-t2; // F
printf('	 delt1 is : %.0f F 
',delt1);
printf('	 delt2 is : %.0f F 
',delt2);
LMTD=((delt2-delt1)/((2.3)*(log10(delt2/delt1))));
printf('	 LMTD is :%.0f F 
',LMTD);
printf('	 On the assumption that the fluids are mixed between passes, each pass must be solved independently. Since only two passes are present in this exchanger, it is simply a matter of assuming the temperature at the end of the first pass. More than half the heat load must be transferred in the first pass; therefore assume ti at the end of the first pass is 125째F 
');
ti=125; // F
tc=((t1)+(ti))/2; // caloric temperature of cold fluid,F
printf('	 caloric temperature of cold fluid is : %.0f F 
',tc);
printf('	 hot fluid:shell side,steam 
');
ho=(1500); // condensation of steam Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',ho);
printf('	 cold fluid:inner tube side,crude 
');
Nt=86;
n=2; // number of passes
L=12; //ft
at1=0.594; // flow area, in^2,from table 10
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=(W/(.177)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gt);
mu2=2.95*2.42; // at 145F,lb/(ft)*(hr)
D=(0.87/12); // ft
Ret1=((D)*(Gt)/mu2); // reynolds number
printf('	 reynolds number is : %.0f 
',Ret1);
mu3=4.8*2.42; // at 110F,lb/(ft)*(hr)
D=(0.87/12); // ft
Ret2=((D)*(Gt)/mu3); // reynolds number
printf('	 reynolds number is : %.0f 
',Ret2);
c=0.485; // Btu/(lb)*(F),at 120F,from fig.2
k=0.0775; // Btu/(hr)*(ft^2)*(F/ft), from table 4
Pr=((c)*(mu3)/k); // prandelt number
printf('	 prandelt number is : %.1f 
',Pr);
Hi=((1.86)*(k/D)*((Ret2*(D/L)*Pr)^(1/3))); // using eq.6.1,Btu/(hr)*(ft^2)*(F)
printf('	 Hi is : %.1f Btu/(hr)*(ft^2)*(F) 
',Hi);
muw=1.2*2.42; // lb/(ft)*(hr),at 249F from fig.14
phyt=(mu3/muw)^0.14;
printf('	 phyt is : %.1f 
',phyt); // from fig.24
hi=(Hi)*(phyt); // from eq.6.37
printf('	 Correct hi to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',hi);
tp=(tc)+(((ho)/(hi+ho))*(T1-tc)); // from eq.5.31
printf('	 tp is : %.0f F 
',tp);
delt=tp-tc; //F
printf('	 delt is : %.0f F 
',delt);
Ai1=0.228 // internal surface per foot of length,ft
Ai=(Nt*L*Ai1/2); // ft^2
printf('	 total surface area is : %.1f ft^2 
',Ai);
delt3=((hi*Ai*delt)/(W*c)); // delt3=ti-t1, F
printf('	 delt3 is : %.1f F 
',delt3);
ti=t1+delt3; // F
printf('	 ti is : %.1f F 
',ti);
printf('	 The oil now enters the second pass at 126.9째F 
');
// end

```

## Example 10.2: Kerosene_Heater.sce

```scilab
printf('	 example 10.2 
');
printf('	 approximate values are mentioned in the book 
');
T1=250; // inlet hot fluid,F
T2=250; // outlet hot fluid,F
t1=95; // inlet cold fluid,F
t2=145; // outlet cold fluid,F
W=16000; // lb/hr
w=423; // lb/hr
printf('	 1.for heat balance 
');
printf('	 for kerosene 
');
c=0.5; // Btu/(lb)*(F)
Q=((W)*(c)*(t2-t1)); // Btu/hr
printf('	 total heat required for kerosene is : %.0f Btu/hr 
',Q);
printf('	 for steam 
');
l=945.5; // Btu/(lb)
Q=((w)*(l)); // Btu/hr
printf('	 total heat required for steam is : %.2e Btu/hr 
',Q);
delt1=T2-t1; //F
delt2=T1-t2; // F
printf('	 delt1 is : %.0f F 
',delt1);
printf('	 delt2 is : %.0f F 
',delt2);
LMTD=((delt2-delt1)/((2.3)*(log10(delt2/delt1))));
printf('	 LMTD is :%.0f F 
',LMTD);
tc=((t1)+(t2))/2; // caloric temperature of cold fluid,F
printf('	 caloric temperature of cold fluid is : %.0f F 
',tc);
printf('	 hot fluid:shell side,steam 
');
ho=(1500); // condensation of steam Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',ho);
printf('	 cold fluid:inner tube side,kerosene 
');
Nt=86;
n=2; // number of passes
L=12; //ft
at1=0.594; // flow area, in^2,from table 10
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=(W/(.177)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gt);
mu2=1.36*2.42; // at 145F,lb/(ft)*(hr)
D=(0.87/12); // ft
Ret1=((D)*(Gt)/mu2); // reynolds number
printf('	 reynolds number is : %.0f 
',Ret1);
mu3=1.75*2.42; // at 120F,lb/(ft)*(hr)
D=(0.87/12); // ft
Ret2=((D)*(Gt)/mu3); // reynolds number
printf('	 reynolds number is : %.1e 
',Ret2);
Z1=331; // Z1=(L*n/D)
jH=3.1; // from fig 24
mu4=1.75; // cp and 40 API
Z2=0.24; // Z2=((k)*(c*mu4/k)^(1/3)), from fig 16
Hi=((jH)*(1/D)*(Z2)); // using eq.6.15a,Btu/(hr)*(ft^2)*(F)
printf('	 Hi is : %.2f Btu/(hr)*(ft^2)*(F) 
',Hi);
ID=0.87; // ft
OD=1; //ft
Hio=(Hi*(ID/OD)); //Btu/(hr)*(ft^2)*(F), from eq.6.5
printf('	 Hio is : %.2f Btu/(hr)*(ft^2)*(F) 
',Hio);
tw=(tc)+(((ho)/(Hio+ho))*(T1-tc)); // from eq.5.31
printf('	 tw is : %.0f F 
',tw);
muw=1.45; // lb/(ft)*(hr),at 249F from fig.14
phyt=(mu3/muw)^0.14;
printf('	 phyt is : %.1f 
',phyt); // from fig.24
hio=(Hio)*(phyt); // from eq.6.37
printf('	 Correct hio to the surface at the OD is : %.1f Btu/(hr)*(ft^2)*(F) 
',hio);
delt=tw-tc; //F
printf('	 delt is : %.0f F 
',delt);
printf('	 Since the kerosene has a viscosity of only 1.75 cp at the caloric temperature and delt=129F, free convection should be investigated. 
');
s=0.8;
row=50; // lb/ft^3, from fig 6
s1=0.810; // at 95F
s2=0.792; // at 145F
bita=((s1^2-s2^2)/(2*(t2-t1)*s1*s2)); // /F
printf('	 beta is : %.6f /F 
',bita);
G=((D^3)*(row^2)*(bita)*(delt)*(4.18*10^8)/(mu3^2));
printf('	 G is : %.1e  
',G);
psy=((2.25)*(1+(0.01*G^(1/3)))/(log10(Ret2)));
printf('	 psy is : %.2f  
',psy);
hio1=(hio*psy);
printf('	 corrected hio1 is : %.1f Btu/(hr)*(ft^2)*(F) 
',hio1);
Uc=((hio1)*(ho)/(hio1+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',Uc);
A2=0.2618; // actual surface supplied for each tube,ft^2,from table 10
A=(Nt*L*A2); // ft^2
printf('	 total surface area is : %.0f ft^2 
',A);
UD=((Q)/((A)*(delt)));
printf('	 actual design overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UD);
Rd=((Uc-UD)/((UD)*(Uc))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.2f (hr)*(ft^2)*(F)/Btu 
',Rd);
// end
```

## Example 10.3: Gas_Oil_Heater_Using_Cores.sce

```scilab
printf('	 example 10.3 
');
printf('	 approximate values are mentioned in the book 
');
T1=250; // inlet hot fluid,F
T2=250; // outlet hot fluid,F
t1=105; // inlet cold fluid,F
t2=130; // outlet cold fluid,F
w=50000; // lb/hr
W=622; // lb/hr
printf('	 1.for heat balance 
');
printf('	 for gas oil 
');
c=0.47; // Btu/(lb)*(F)
Q=((w)*(c)*(t2-t1)); // Btu/hr
printf('	 total heat required for gas oil is : %.2e Btu/hr 
',Q);
printf('	 for steam 
');
l=945.5; // Btu/(lb)
Q=((W)*(l)); // Btu/hr
printf('	 total heat required for steam is : %.2e Btu/hr 
',Q);
delt1=T2-t1; //F
delt2=T1-t2; // F
printf('	 delt1 is : %.0f F 
',delt1);
printf('	 delt2 is : %.0f F 
',delt2);
LMTD=((delt2-delt1)/((2.3)*(log10(delt2/delt1))));
printf('	 LMTD is :%.1f F 
',LMTD);
tc=((t1)+(t2))/2; // caloric temperature of cold fluid,F
printf('	 caloric temperature of cold fluid is : %.1f F 
',tc);
printf('	 hot fluid:shell side,steam 
');
ID=15.25; // in
C=0.25; // clearance
B=15; // baffle spacing,in
PT=1.25;
as=((ID*C*B)/(144*PT)); // flow area,ft^2, eq 7.1
printf('	 flow area is : %.3f ft^2 
',as);
Gs=(6220/as); // mass velocity,lb/(hr)*(ft^2), calculation mistake
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gs);
mu1=0.0314; // at 250F,lb/(ft)*(hr), from fig.15
De=0.060; // from fig.29,ft
Res=((De)*(Gs)/mu1); // reynolds number, calculation mistake
printf('	 reynolds number is : %.2e 
',Res);
ho=1500; //Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',ho);
printf('	 cold fluid:inner tube side,crude oil 
');
d1=0.5; // in
d2=0.87; // in
at1=((3.14*(d2^2-d1^2))/4);
printf('	 at1 is : %.1f in^2 
',at1);
Nt=86;
n=2; // number of passes
L=12; //ft
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=(w/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.1e lb/(hr)*(ft^2) 
',Gt);
De=(d2^2-d1^2)/(12*d2);
printf('	 De is : %.4f ft 
',De);
mu2=16.7; // at 117F,lb/(ft)*(hr)
Ret=((De)*(Gt)/mu2); // reynolds number
printf('	 reynolds number is : %.2e 
',Ret);
jH=3.1; // from fig.24
Z=0.35; // Z=(K*(c*mu3/k)^(1/3)),Btu/(hr)(ft^2)(F/ft), at mu3=6.9cp and 28 API
Hi=((jH)*(1/De)*(Z)); //Hi=(hi/phyp),using eq.6.15a,Btu/(hr)*(ft^2)*(F)
printf('	 Hi is : %.1f Btu/(hr)*(ft^2)*(F) 
',Hi);
ID=0.87; // ft
OD=1; //ft
Hio=((Hi)*(ID/OD)); //Hio=(hio/phyp), using eq.6.5
printf('	 Correct Hi0 to the surface at the OD is : %.1f Btu/(hr)*(ft^2)*(F) 
',Hio);
muw=4.84; // lb/(ft)*(hr), from fig.14
phyt=(mu2/muw)^0.14;
printf('	 phyt is : %.2f 
',phyt); // from fig.24
hio=(Hio)*(phyt); // from eq.6.37
printf('	 Correct hi0 to the surface at the OD is : %.1f Btu/(hr)*(ft^2)*(F) 
',hio);
tw=(tc)+(((ho)/(hio+ho))*(T1-tc)); // from eq.5.31
printf('	 tw is : %.0f F 
',tw);
Uc=((hio)*(ho)/(hio+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',Uc);
A=270; // ft^2
printf('	 total surface area is : %.0f ft^2 
',A);
UD=((Q)/((A)*(LMTD)));
printf('	 actual design overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UD);
Rd=((Uc-UD)/((UD)*(Uc))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.3f (hr)*(ft^2)*(F)/Btu 
',Rd);
printf('	 pressure drop  for annulus 
');
f=0.0016; // friction factor for reynolds number 25300, using fig.29
s=0.00116; // for reynolds number 25300,using fig.6
Ds=15.25/12; // ft
phys=1;
N=(12*L/B); // number of crosses,using eq.7.43
printf('	 number of crosses are : %.0f 
',N);
delPs=((f*(19600^2)*(Ds)*(N))/(5.22*(10^10)*(De)*(s)*(phys)))/(2); // using eq.7.44,psi
printf('	 delPs is : %.1f psi 
',delPs);
printf('	 pressure drop  for inner pipe 
');
dt=(d2-d1)/(12); // ft
printf('	 dt is : %.4f ft 
',dt);
Ret2=(dt*Gt/mu2);
printf('	 Ret2 is : %.0f 
',Ret2);
f=0.00066; // friction factor for reynolds number 8220, using fig.26
phyt=1.35; // fig 6
printf('	 phyt is : %.2f 
',phyt);
s=0.85;
delPt=((f*(420000^2)*(L)*(n))/(5.22*(10^10)*(0.0309)*(s)*(phyt))); // using eq.7.45,psi
printf('	 delPt is : %.1f psi 
',delPt);
printf('	 delPr is negligible 
');
//end
```

## Example 10.4: Calculation_of_a_Heating_Bundle_for_an_Aniline_Storage_Tank.sce

```scilab
printf('	 example 10.4 
');
printf('	 approximate values are mentioned in the book 
');
t1=100; // F
t2=0; // F
T1abs=100+460; // R
T2abs=460; //R
delt=t1-t2;
printf('	 delt is : %.f F 
',delt);
hc=0.3*(delt^0.25); // convection loss, Btu/(hr)*(ft^2)*(째F)
printf('	 convection loss is : %.2f Btu/(hr)(ft^2)(F) 
',hc);
e=0.8; // emissivity
hr=((0.173*e*((T1abs/100)^4-(T2abs/100)^4))/(T1abs-T2abs)); // radiation rate, from 4.32, Btu/(hr)(ft^2)(F)
printf('	 radiation loss is : %.2f Btu/(hr)(ft^2)(F) 
',hr);
hl=hc+hr; // combined loss, Btu/(hr)(ft^2)(F)
printf('	 combined loss is : %.1f Btu/(hr)(ft^2)(F) 
',hl);
D=5; // ft
L=12; // ft
A1=((2*3.14*D^2)/(4))+(3.14*D*L); // total tank area
printf('	 total tank area is : %.1f ft^2 
',A1);
Q=(hl*A1*delt); // total heat loss
printf('	 total heat loss : %.2e Btu/hr 
',Q);
printf('	 This heat must be supplied by the pipe bundle,Assuming exhaust steam to be at 212째F 
');
d0=1.32;
X=(delt/d0);
tf=((t1+212)/2); // F
printf('	 X is : %.0f 
',X);
printf('	 tf is : %.0f F 
',tf);
hio=48; // from fig 10.4, Btu/(hr)(ft^2)(F)
ho=1500; // condensation of steam,Btu/(hr)(ft^2)(F)
Uc=((hio)*(ho)/(hio+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',Uc);
Rd=0.02; // dirt factor, (hr)(ft^2)(F)/Btu
UD=((Uc)/((1)+(Uc*Rd))); // design overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 design overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UD);
A2=((Q)/((UD)*(212-100))); // total surface,ft^2
printf('	 total surface is : %.1f ft^2 
',A2);
A3=2.06; // area/pipe
N=(A2/A3);
printf('	 number of pipes are : %.0f 
',N);
//end
```
