## Example 12.1: Calculation_of_a_Horizontal_n_Propanol_Condenser.sce

```scilab
printf('	 example 12.1 
');
printf('	 approximate values are mentioned in the book 
');
T1=244; // inlet hot fluid,F
T2=244; // outlet hot fluid,F
t1=85; // inlet cold fluid,F
t2=120; // outlet cold fluid,F
W=60000; // lb/hr
w=488000; // lb/hr
printf('	 1.for heat balance 
');
printf('	 for propanol 
');
l=285; // Btu/(lb)
Q=((W)*(l)); // Btu/hr
printf('	 total heat required for propanol is : %.2e Btu/hr 
',Q);
printf('	 for water 
');
c=1; // Btu/(lb)*(F)
Q=((w)*(c)*(t2-t1)); // Btu/hr
printf('	 total heat required for water is : %.2e Btu/hr 
',Q);
delt1=T2-t1; //F
delt2=T1-t2; // F
printf('	 delt1 is : %.0f F 
',delt1);
printf('	 delt2 is : %.0f F 
',delt2);
LMTD=((delt2-delt1)/((2.3)*(log10(delt2/delt1))));
printf('	 LMTD is :%.0f F 
',LMTD);
Tc=((T2)+(T1))/(2); // caloric temperature of hot fluid,F
printf('	 caloric temperature of hot fluid is : %.0f F 
',Tc);
tc=((t1)+(t2))/(2); // caloric temperature of cold fluid,F
printf('	 caloric temperature of cold fluid is : %.1f F 
',tc);
UD1=100; // assume, from table 8
A1=((Q)/((UD1)*(LMTD)));
printf('	 A1 is : %.0f ft^2 
',A1);
a1=0.1963; // ft^2/lin ft
N1=(A1/(8*a1));
printf('	 number of tubes are : %.0f 
',N1);
N2=766; // assuming 4 tube passes, from table 9
A2=(N2*8*a1); // ft^2
printf('	 total surface area is : %.0f ft^2 
',A2);
UD=((Q)/((A2)*(LMTD)));
printf('	 correct design overall coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',UD);
printf('	 hot fluid:shell side,propanol 
');
ID=31; // in
C=0.1875; // clearance
B=31; // baffle spacing,in
PT=0.937;
L=8; // ft
as=((ID*C*B)/(144*PT)); // flow area,from eq 7.1,ft^2
printf('	 flow area is : %.2f ft^2 
',as);
Gs=(W/as); // mass velocity,from eq 7.2,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gs);
G1=(W/(L*N2^(2/3))); // from eq.12.43
printf('	 G1 is : %.1f lb/(hr)*(lin ft) 
',G1);
printf('	 cold fluid:inner tube side,water 
');
Nt=766;
n=4; // number of passes
L=8; //ft
at1=0.302; // flow area, in^2
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=(w/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gt);
V=(Gt/(3600*62.5));
printf('	 V is : %.2f fps 
',V);
mu2=1.74; // at 102.5F,lb/(ft)*(hr)
D=0.0517; // ft
Ret=((D)*(Gt)/mu2); // reynolds number
printf('	 reynolds number is : %.2e 
',Ret);
hi=1300; //Btu/(hr)*(ft^2)*(F)
printf('	 hi is : %.0f Btu/(hr)*(ft^2)*(F) 
',hi);
ID=0.62; // ft
OD=0.75; //ft
hio=((hi)*(ID/OD)); // using eq.6.5
printf('	 Correct hi0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',hio); // calculation mistake
ho=200; // assumption
tw=(tc)+(((ho)/(hio+ho))*(Tc-tc)); // from eq.5.31
printf('	 tw is : %.0f F 
',tw);
tf=(Tc+tw)/(2); // from eq 12.19
printf('	 tf is : %.1f F 
',tf);
kf=0.094; // Btu/(hr)*(ft^2)*(F/ft), from table 4
sf=0.8; // from table 6
muf=0.62; // cp, from fig 14
ho=172; // Btu/(hr)*(ft^2)*(F), from fig 12.9
printf('	 Correct ho to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',ho);
printf('	 Based on h=172 instead of the assumed 200 a new value of tw,and tf could be obtained to give a more exact value of h based on fluid properties at a value of tf more nearly correct 
');
printf('	 pressure drop  for annulus 
');
mu1=0.0242; // lb/(ft)*(hr), fir 15
De=0.0458; // fig 28
Res=((De)*(Gs)/mu1); // reynolds number
printf('	 reynolds number is : %.2e 
',Res);
f=0.00141; // friction factor for reynolds number 84600, using fig.29
s=0.00381; // for reynolds number 84600,using fig.6
Ds=31/12; // ft
phys=1;
N=(3); // number of crosses,using eq.7.43
printf('	 number of crosses are : %.0f 
',N);
delPs=((f*(Gs^2)*(Ds)*(N))/(5.22*(10^10)*(De)*(s)*(phys)))/(2); // using eq.12.47,psi
printf('	 delPs is : %.1f psi 
',delPs);
printf('	 allowable delPa is 2 psi 
');
printf('	 pressure drop  for inner pipe 
');
f=0.00019; // friction factor for reynolds number 36200, using fig.26
s=1;
phyt=1;
delPt=((f*(Gt^2)*(L)*(n))/(5.22*(10^10)*(D)*(s)*(phyt))); // using eq.7.45,psi
printf('	 delPt is : %.1f psi 
',delPt);
X1=0.2; // X1=((V^2)/(2*g)),using fig.27
delPr=((4*n*X1)/(s)); // using eq.7.46,psi
printf('	 delPr is : %.1f psi 
',delPr);
delPT=delPt+delPr; // using eq.7.47,psi
printf('	 delPT is : %.1f psi 
',delPT);
printf('	 allowable delPT is 10 psi 
');
Uc=((hio)*(ho)/(hio+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',Uc);
Rd=((Uc-UD)/((UD)*(Uc))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.4f (hr)*(ft^2)*(F)/Btu 
',Rd);
// end
```

## Example 12.3: Calculation_of_a_Butane_Desuperheater_condenser.sce

```scilab
printf('	 example 12.3 
');
printf('	 approximate values are mentioned in the book 
');
T1=200; // inlet hot fluid,F
T2=130; // outlet hot fluid,F
T3=125; // after condensation
t1=65; // inlet cold fluid,F
t3=100; // outlet cold fluid,F
W=27958; // lb/hr
w=135500; // lb/hr
printf('	 1.for heat balance 
');
printf('	 for butane 
');
c=0.44; // Btu/(lb)(F)
qd=((W)*(c)*(T1-T2)); // Btu/hr
printf('	 total heat required for desuperheating of butane is : %.1e Btu/hr 
',qd);
HT2=309; // enthalpy at T2, Btu/lb
HT3=170; // enthalpy at T3, Btu/lb
qc=(W*(HT2-HT3)); // for condensation
printf('	 total heat required for condensing of butane is : %.2e Btu/hr 
',qc);
Q=qd+qc;
printf('	 total heat required for  butane is : %.2e Btu/hr 
',Q);
printf('	 for water 
');
c=1; // Btu/(lb)*(F)
Q=((w)*(c)*(t3-t1)); // Btu/hr
printf('	 total heat required for water is : %.2e Btu/hr 
',Q);
deltw=(qc/w);
printf('	 deltw is : %.1f F 
',deltw);
t2=t1+deltw;
printf('	 t2 is : %.1f F 
',t2)
printf('	 for desuperheating 
');
delt1=T2-t2; //F
delt2=T1-t3; // F
printf('	 delt1 is : %.0f F 
',delt1);
printf('	 delt2 is : %.0f F 
',delt2);
LMTDd=((delt2-delt1)/((2.3)*(log10(delt2/delt1))));
printf('	 LMTD is :%.0f F 
',LMTDd);
w1=(qd/LMTDd);
printf('	 w1 is : %.3e lb/hr 
',w1);
printf('	 for condensing 
');
delt3=T3-t1; //F
delt4=T2-t2; // F
printf('	 delt1 is : %.0f F 
',delt3);
printf('	 delt2 is : %.0f F 
',delt4);
LMTDc=((delt4-delt3)/((2.3)*(log10(delt4/delt3))));
printf('	 LMTD is :%.0f F 
',LMTDc);
w2=(qc/LMTDc);
printf('	 w1 is : %.2e lb/hr 
',w2);
delt=(Q/(w1+w2));
printf('	 delt is : % .1f F 
',delt);
Tc=((T3)+(T2))/(2); // caloric temperature of hot fluid,F
printf('	 caloric temperature of hot fluid is : %.1f F 
',Tc);
tc=((t1)+(t3))/(2); // caloric temperature of cold fluid,F
printf('	 caloric temperature of cold fluid is : %.1f F 
',tc);
printf('	 hot fluid:shell side,butane 
');
ID=23.25; // in
C=0.25; // clearance
B=12; // baffle spacing,in
PT=1;
as=((ID*C*B)/(144*PT)); // flow area,ft^2
printf('	 flow area is : %.3f ft^2 
',as);
printf('	 desuperheating 
');
Gs=(W/as); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gs);
mu1=0.0242; // at 165F,lb/(ft)*(hr), from fig.15
De=0.73/12; // from fig.28,ft
Res=((De)*(Gs)/mu1); // reynolds number
printf('	 reynolds number is : %.2e 
',Res);
jH=239; // from fig.28
k=0.012; // Btu/(hr)*(ft^2)*(F/ft), from table 5
Z=0.96; // Z=((c)*(mu1)/k)^(1/3)
ho=((jH)*(k/De)*(Z)); // H0=(h0/phya),using eq.6.15b,Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',ho);
printf('	 cold fluid:inner tube side,water 
');
Nt=352;
n=4; // number of passes
L=16; //ft
at1=0.302; // flow area,table 10, in^2
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=(w/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gt);
V=(Gt/(3600*62.5));
printf('	 V is : %.2f fps 
',V);
mu2=2.11; // at 82.5F, fig 14,lb/(ft)*(hr)
D=0.0517; // ft
Ret=((D)*(Gt)/mu2); // reynolds number
printf('	 reynolds number is : %.2e 
',Ret);
hi=800; // fig 25,Btu/(hr)*(ft^2)*(F)
printf('	 hi is : %.0f Btu/(hr)*(ft^2)*(F) 
',hi);
ID=0.62; // ft
OD=0.75; //ft
hio=((hi)*(ID/OD)); // using eq.6.5
printf('	 Correct hio to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',hio);
Ud=((hio)*(ho)/(hio+ho)); // clean overall coefficient,eq 6.38,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',Ud);
Ad=(qd/(Ud*LMTDd));
printf('	 clean surface required for desuperheating : %.0f ft^2 
',Ad);
printf('	 for condensaton 
');
Lc=16*0.6; // condensation occurs 60% of the tube length
G1=(W/(Lc*Nt^(2/3))); // from eq.12.43
printf('	 G1 is : %.1f lb/(hr)*(lin ft) 
',G1);
ho=200; // assumption
tw=(tc)+(((ho)/(hio+ho))*(Tc-tc)); // from eq.5.31
printf('	 tw is : %.0f F 
',tw);
tf=(Tc+tw)/(2); // from eq 12.19
printf('	 tf is : %.0f F 
',tf);
kf=0.075; // Btu/(hr)*(ft^2)*(F/ft)
sf=0.55; // from table 6
muf=0.14; // cp, from fig 14
ho=207; // Btu/(hr)*(ft^2)*(F), from fig 12.9
printf('	 Correct ho to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',ho);
Uc=((hio)*(ho)/(hio+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Uc);
Ac=(qc/(Uc*LMTDc));
printf('	 clean surface required for desuperheating : %.0f ft^2 
',Ac);
AC=Ad+Ac;
printf('	 total clean surface : %.0f ft^2 
',AC);
lc=(Ac/(Ac+Ad));
printf('	 assumed condensing length percentage : %.2f 
',lc);
UC=((Ud*Ad)+(Uc*Ac))/(AC);
printf('	 weighted clean overall coefficient : %.0f Btu/(hr)*(ft^2)*(F) 
',UC);
A2=0.1963; // actual surface supplied for each tube,ft^2,from table 10
A=(Nt*L*A2); // ft^2
printf('	 total surface area is : %.0f ft^2 
',A);
UD=((Q)/((A)*(delt)));
printf('	 actual design overall coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',UD);
Rd=((UC-UD)/((UD)*(UC))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.4f (hr)*(ft^2)*(F)/Btu 
',Rd);
printf('	 pressure drop  for annulus 
');
printf('	 desuperheating 
');
Ld=6.4; //ft
De=0.0608; // fig 28
f=0.0013; // friction factor for reynolds number 145000, using fig.29
Ds=1.94; // ft
phys=1;
N=(12*Ld/B); // number of crosses,using eq.7.43
printf('	 number of crosses are : %.0f 
',N);
row=(58.1/((359)*(625/492)*(14.7/99.7)));
printf('	 row is %.3f lb/ft^3 
',row);
s=(row/62.5);
printf('	 s is %.4f 
',s);
delPsd=((f*(Gs^2)*(Ds)*(N))/(5.22*(10^10)*(De)*(s)*(phys))); // using eq.7.44,psi
printf('	 delPs is : %.1f psi 
',delPsd);
printf('	 condensation 
');
N=(12*Lc/B); // number of crosses,using eq.7.43
printf('	 number of crosses are : %.0f 
',N);
delPsc=((f*(Gs^2)*(Ds)*(N))/(5.22*(10^10)*(De)*(s)*(phys)))/(2); // using eq 12.47,psi
printf('	 delPsc is : %.1f psi 
',delPsc);
delPS=delPsd+delPsc;
printf('	 delPS is : %.0f psi 
',delPS);
printf('	 allowable delPa is 2 psi 
');
printf('	 pressure drop  for inner pipe 
');
f=0.00023; // friction factor for reynolds number 17900, using fig.26
s=1;
phyt=1;
delPt=((f*(Gt^2)*(L)*(n))/(5.22*(10^10)*(D)*(s)*(phyt))); // using eq.7.45,psi
printf('	 delPt is : %.0f psi 
',delPt);
X1=0.075; // X1=((V^2)/(2*g)),using fig.27
delPr=((4*n*X1)/(s)); // using eq.7.46,psi
printf('	 delPr is : %.1f psi 
',delPr);
delPT=delPt+delPr; // using eq.7.47,psi
printf('	 delPT is : %.1f psi 
',delPT);
printf('	 allowable delPa is 10 psi 
');
//end
```
