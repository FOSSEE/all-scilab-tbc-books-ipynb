## Example 15.1: Calculation_of_the_Average_Specific_Volume.sce

```scilab
printf('	 example 15.1 
');
printf('	 approximate values are mentioned in the book 
');
ts=250;
T1=400;
T2=300;
w=10000; // lb/hr
W=150000; // lb/hr
l=945.3; // Btu/(lb) , table 7
Q=((w)*(l)); // Btu/hr
printf('	 total heat required for steam is : %.2e Btu/hr 
',Q);
C=0.63; // Btu/(lb)*(F)
Q=((W)*(C)*(T1-T2)); // Btu/hr
printf('	 total heat required for kerosene is : %.2e Btu/hr 
',Q); 
delt1=T2-ts; //F
delt2=T1-ts; // F
printf('	 delt1 is : %.0f F 
',delt1);
printf('	 delt2 is : %.0f F 
',delt2);
LMTD=((delt2-delt1)/((2.3)*(log10(delt2/delt1))));
printf('	 LMTD is :%.0f F 
',LMTD);
UD=100;
A=(Q/(UD*LMTD));
printf('	 A : %.2e ft^2 
',A);
WC=94500; // Btu/F
vl=0.017; // ft^3/lb, from table 7
vv=13.75; // ft^3/lb, from table 7
printf('	 By the law of mixtures 
');
// Assume 80 per cent of the outlet fluid is vapor
v2=(0.8*vv)+(.2*vl);
printf('	 v2 : %.0f ft^3/lb 
',v2);
vav=(WC*(v2-vl)/(UD*A))-((WC*(T2-ts)/(l*w))*(vv-vl))+vl;
printf('	 vav : %.2f ft^3/lb 
',vav);
printf('	 By the approximate method 
');
vav1=(vl+v2)/(2);
printf('	 vav : %.2f ft^3/lb 
',vav1);
row=62.5;
rowac=(1/vav);
s=(rowac/row);
printf('	 actual density : %.3f lb/ft^3 
',rowac);
printf('	 s : %.4f 
',s);
rowap=(1/vav1);
s=(rowap/row);
printf('	 approximate density : %.3f lb/ft^3 
',rowac);
printf('	 s : %.4f 
',s);
// end
```

## Example 15.2: Vaporizer_or_Pump_through_Reboiler_with_Isothermal_Boiling.sce

```scilab
printf('	 example 15.2 
');
printf('	 approximate values are mentioned in the book 
');
t1=108; // inlet cold fluid,F
t2=235; // outlet cold fluid,F
Ts=338;
Wp=24700; // lb/hr
Wv=19750; // lb/hr
w=4880; // lb/hr
printf('	 1.for heat balance 
');
Ht1=162; // enthalpy at t1, Btu/lb, fig 9
Ht2=248; // enthalpy at t2, Btu/lb, fig 9
qp=(Wp*(Ht2-Ht1)); // for preheat
printf('	 total heat required for preheat of butane is : %.2e Btu/hr 
',qp);
Ht3=358; // enthalpy of vapour at t2, Btu/lb, fig 9
qv=Wv*(Ht3-Ht2);
printf('	 total heat required for vapourisation of butane is : %.2e Btu/hr 
',qv);
Q=qp+qv;
printf('	 total heat required for  butane is : %.2e Btu/hr 
',Q);
printf('	 for steam 
');
l=880.6; // Btu/(lb), table 7
Q=((w)*(l)); // Btu/hr
printf('	 total heat required for steam is : %.2e Btu/hr 
',Q);
deltp=158.5; // F, from eq 5.14
deltv=103; // F eq 5.14
Wp1=(qp/deltp);
printf('	 Wp1 is : %.2e lb/hr 
',Wp1);
Wv1=(qv/deltv);
printf('	 Wv1 is : %.2e lb/hr 
',Wv1);
W=(Wp1+Wv1);
printf('	 W is : %.2e lb/hr 
',W);
delt=(Q/W);
printf('	 weighted delt is : % .1f F 
',delt);
Tc=((Ts)+(Ts))/(2); // caloric temperature of hot fluid,F
printf('	 caloric temperature of hot fluid is : %.1f F 
',Tc);
tc=((t1)+(t2))/(2); // caloric temperature of cold fluid,F
printf('	 caloric temperature of cold fluid is : %.1f F 
',tc);
printf('	 hot fluid:inner tube side,steam 
');
Nt=76;
n=2; // number of passes
L=16; //ft
at1=0.594; // flow area,table 10, in^2
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=(w/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gt);
mu1=0.0363; // at 338F, fig 15,lb/(ft)*(hr)
D=0.0725; // ft
Ret=((D)*(Gt)/mu1); // reynolds number
printf('	 reynolds number is : %.1e 
',Ret);
hio=1500; // condensing steam,Btu/(hr)*(ft^2)*(F)
printf('	 hio is : %.0f Btu/(hr)*(ft^2)*(F) 
',hio);
printf('	 cold fluid:shell side,butane 
');
printf('	 preheating 
');
ID=15.25; // in
C=0.25; // clearance
B=5; // baffle spacing,in
PT=1.25;
as=((ID*C*B)/(144*PT)); // flow area,ft^2
printf('	 flow area is : %.3f ft^2 
',as);
Gs=(Wp/as); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gs);
mu2=0.278; // at 172F,lb/(ft)*(hr), from fig.14
De=0.0825; // from fig.28,ft
Res=((De)*(Gs)/mu2); // reynolds number
printf('	 reynolds number is : %.2e 
',Res);
jH=159; // from fig.28
Z=0.12; // Z=k*((c)*(mu1)/k)^(1/3), fig 16
hop=((jH)*(1/De)*(Z)); //using eq.6.15b,Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',hop);
Up=((hio)*(hop)/(hio+hop)); // clean overall coefficient,eq 6.38,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient for preheating : %.0e Btu/(hr)*(ft^2)*(F) 
',Up);
Ap=(qp/(Up*deltp));
printf('	 clean surface required for preheating : %.0f ft^2 
',Ap);
printf('	 for vapourisation 
');
mu2=0.242; // at 172F,lb/(ft)*(hr), from fig.14
Res=((De)*(Gs)/mu2); // reynolds number
printf('	 reynolds number is : %.2e 
',Res);
jH=170; // from fig.28
Z=0.115; // Z=k*((c)*(mu1)/k)^(1/3), fig 16
hov=((jH)*(1/De)*(Z)); //using eq.6.15b,Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',hov);
Uv=((hio)*(hov)/(hio+hov)); // clean overall coefficient,eq 6.38,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient for vapourisation : %.0f Btu/(hr)*(ft^2)*(F) 
',Uv);
Av=(qv/(Uv*deltv));
printf('	 clean surface required for vapourisation : %.0f ft^2 
',Av);
Ac=Ap+Av;
printf('	 total clean surface : %.1e ft^2 
',Ac);
UC=((Up*Ap)+(Uv*Av))/(Ac);
printf('	 weighted clean overall coefficient : %.0f Btu/(hr)*(ft^2)*(F) 
',UC);
A2=0.2618; // actual surface supplied for each tube,ft^2,from table 10
A=(Nt*L*A2); // ft^2
printf('	 total surface area is : %.0f ft^2 
',A);
UD=((Q)/((A)*(delt)));
printf('	 actual design overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UD);
// A total of 170 ft2 are required of which 103 are to be used for vaporization. For the total surface required 318 ft2 will be provided. It can be assumed, then, that the surface provided for vaporization is 193ft^2
// then flux is Q/A=10700, which is with in satisfactory levels.
Rd=((UC-UD)/((UD)*(UC))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.4f (hr)*(ft^2)*(F)/Btu 
',Rd);
printf('	 pressure drop  for inner pipe 
');
f=0.000165; // friction factor for reynolds number 62000, using fig.26
s=0.00413;
phyt=1;
delPt=((f*(Gt^2)*(L)*(n))/(5.22*(10^10)*(D)*(s)*(phyt)))/(2); // using eq.7.45,psi
printf('	 delPt is : %.2f psi 
',delPt);
printf('	 allowable delPa is negligible 
');
printf('	 pressure drop  for annulus 
');
printf('	 preheating 
');
f=0.00145; // friction factor for reynolds number 69200, using fig.29
Lp=(L*Ap/Ac); //ft
printf('	 length of preheat zone : %.1f ft 
',Lp);
N=(12*Lp/B); // number of crosses,using eq.7.43
printf('	 number of crosses are : %.0f 
',N);
s=0.5; // for reynolds number 69200,using fig.6
Ds=1.27; // fig 28
phys=1;
delPsp=((f*(Gs^2)*(Ds)*(N))/(5.22*(10^10)*(De)*(s)*(phys))); // using eq.7.44,psi
printf('	 delPsp is : %.1f psi 
',delPsp);
printf('	 vapourisation 
');
f=0.00142;
Lv=9.7; // Lv=L-Lp
Nv=(12*Lv/B); // number of crosses,using eq.7.43
printf('	 number of crosses are : %.0f 
',Nv);
s=0.28; 
delPsv=((f*(Gs^2)*(Ds)*(Nv))/(5.22*(10^10)*(De)*(s)*(1))); // using eq 12.47,psi
printf('	 delPsv is : %.1f psi 
',delPsv);
delPS=delPsp+delPsv;
printf('	 delPS is : %.1f psi 
',delPS);
printf('	 allowable delPa is 5 psi 
');
//end
```

## Example 15.3: Calculation_of_a_Kettle_Reboiler.sce

```scilab
printf('	 example 15.3 
');
printf('	 approximate values are mentioned in the book 
');
ts=400;
T1=575;
T2=475;
W=28100; // lb/hr
w=34700; // lb/hr
printf('	 1.for heat balance 
');
HT1=290; // enthalpy at T1, Btu/lb, fig 11
HT2=385; // enthalpy at T2, Btu/lb, fig 11
Q=(W*(HT2-HT1)); // for preheat
printf('	 total heat required for gasoline is : %.2e Btu/hr 
',Q);
c=0.77; // Btu/(lb), table 7
Q=((w)*(c)*(T1-T2)); // Btu/hr
printf('	 total heat required for gasoil is : %.2e Btu/hr 
',Q);
delt=118; // F eq 5.14
S=((T2-ts)/(T1-ts));
printf('	 S is : %.3f 
',S);
Kc=0.37; // fig 17
Fc=0.42;
Tc=(T2+(0.42*(T1-T2)));
printf('	 Tc is : %.0f F 
',Tc);
printf('	 hot fluid:inner tube side,gasoil 
');
Nt=68;
n=6; // number of passes
L=12; //ft
at1=0.546; // flow area,table 10, in^2
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=(w/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gt);
mu1=0.65; // at 517F, fig 14,lb/(ft)*(hr)
D=0.0694; // ft
Ret=((D)*(Gt)/mu1); // reynolds number
printf('	 reynolds number is : %.2e 
',Ret);
jH=220; // from fig.24
Z=0.118; // Z=k*((c)*(mu1)/k)^(1/3), fig 16
Hi=((jH)*(1/D)*(Z)); //hi/phyt, Hi=()using eq.6.15d,Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Hi);
Hio=((Hi)*(0.834/1)); //Hio=(hio/phyp), using eq.6.9
printf('	 Correct Hi0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',Hio);
// (mu1/muw)^(0.14) is negligible
printf('	 cold fluid:shell side,gasoline 
');
ho=300; // assumption
tw=(ts)+(((Hio)/(Hio+ho))*(Tc-ts)); // from eq.5.31
printf('	 tw is : %.0f F 
',tw);
deltw=(tw-ts);
printf('	 deltw : %.0f F 
',deltw);
// from fig 15.11, ho>300
Uc=((Hio)*(ho)/(Hio+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Uc);
A2=0.2618; // actual surface supplied for each tube,ft^2,from table 10
A=(Nt*L*A2); // ft^2
printf('	 total surface area is : %.0f ft^2 
',A);
UD=((Q)/((A)*(delt)));
printf('	 actual design overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UD);
// check for max. flux=Q/A=12500.(satisfactory)
Rd=((Uc-UD)/((UD)*(Uc))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.4f (hr)*(ft^2)*(F)/Btu 
',Rd);
printf('	 pressure drop  for inner pipe 
');
f=0.00015; // friction factor for reynolds number 85700, using fig.26
s=0.71;
phyt=1;
delPt=((f*(Gt^2)*(L)*(n))/(5.22*(10^10)*(D)*(s)*(phyt))); // using eq.7.45,psi
printf('	 delPt is : %.1f psi 
',delPt);
X1=0.09; // X1=((V^2)/(2*g)), for Gt 1060000,using fig.27
delPr=((4*n*X1)/(s)); // using eq.7.46,psi
printf('	 delPr is : %.1f psi 
',delPr);
delPT=delPt+delPr; // using eq.7.47,psi
printf('	 delPT is : %.1f psi 
',delPT);
printf('	 allowable delPa is 10psi 
');
printf('	 delPs is negligible 
');
//end
```

## Example 15.4: Calculation_of_a_Once_through_Horizontal_Thermosyphon_Reboiler.sce

```scilab
printf('	 example 15.4
');
printf('	 approximate values are mentioned in the book 
');
t1=315; // inlet cold fluid,F
t2=335; // outlet cold fluid,F
T1=525;
T2=400;
Wv=29000; // lb/hr
Ws=38500; // lb/hr
w=51000; // lb/hr
printf('	 1.for heat balance 
');
Ht1=238; // enthalpy at t1, Btu/lb, fig 9
Ht2=252; // enthalpy at t2, Btu/lb, fig 9
Ht3=378; // enthalpy of vapour at t2 
qv=(Wv*(Ht3-Ht2)); // for preheat
printf('	 qv is : %.2e Btu/hr 
',qv);
qs=Ws*(Ht2-Ht1);
printf('	 qs is : %.2e Btu/hr 
',qs);
Q=qs+qv;
printf('	 total heat required for naphtha is : %.2e Btu/hr 
',Q);
c=0.66; // Btu/(lb)(F)
Q=((w)*(c)*(T1-T2)); // Btu/hr
printf('	 total heat required for gasoil is : %.2e Btu/hr 
',Q);
delt1=T2-t1; //F
delt2=T1-t2; // F
printf('	 delt1 is : %.0f F 
',delt1);
printf('	 delt2 is : %.0f F 
',delt2);
LMTD=((delt2-delt1)/((2.3)*(log10(delt2/delt1))));
printf('	 LMTD is :%.0f F 
',LMTD);
R=((T1-T2)/(t2-t1));
printf('	 R is : %.2f 
',R);
S=((t2-t1)/(T1-t1));
printf('	 S is : %.3f 
',S);
printf('	 FT is 0.97 
'); // from fig 18
delt=(0.97*LMTD); // F
printf('	 delt is : %.0f F 
',delt);
X=((delt1)/(delt2)); // fig 17
printf('	 ratio of two local temperature difference is : %.3f 
',X);
Fc=0.41; // from fig.17
Kc=0.42;
Tc=((T2)+((Fc)*(T1-T2))); // caloric temperature of hot fluid,F
printf('	 caloric temperature of hot fluid is : %.0f F 
',Tc);
tc=((t1)+((Fc)*(t2-t1))); // caloric temperature of cold fluid,F
printf('	 caloric temperature of cold fluid is : %.0f F 
',tc);
printf('	 hot fluid:inner tube side,steam 
');
Nt=116;
n=8; // number of passes
L=12; //ft
at1=0.546; // flow area,table 10, in^2
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=(w/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gt);
mu1=1.09; // at 451F, fig 14,lb/(ft)*(hr)
D=0.0695; // ft
Ret=((D)*(Gt)/mu1); // reynolds number
printf('	 reynolds number is : %.2e 
',Ret);
jH=168; // from fig.24
Z=0.142; // Z=k*((c)*(mu1)/k)^(1/3), fig 16
Hi=((jH)*(1/D)*(Z)); //, Hi=(hi/phyt)using eq.6.15d,Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Hi);
Hio=((Hi)*(0.834/1)); //Hio=(hio/phyp), using eq.6.9
printf('	 Correct Hio to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',Hio);
printf('	 cold fluid:shell side,naphtha 
');
ho1=200; // assumption
tw=(tc)+(((Hio)/(Hio+ho1))*(Tc-tc)); // from eq.5.31, calculation mistake
printf('	 tw is : %.0f F 
',tw);
deltw=(tw-tc);
printf('	 deltw : %.0f F 
',deltw);
// from fig 15.11, hv>300, hs=60
Av=(qv/300);
As=qs/60;
printf('	 qv/hv : %.3e 
',Av);
printf('	 qs/hs : %.0e 
',As);
A1=As+Av;
printf('	 A : %.3e 
',A1);
ho=(Q/A1);
printf('	 ho : %.0f 
',ho);
Uc=((Hio)*(ho)/(Hio+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Uc);
A2=0.2618; // actual surface supplied for each tube,ft^2,from table 10
A=(Nt*L*A2); // ft^2
printf('	 total surface area is : %.0f ft^2 
',A);
UD=((Q)/((A)*(delt)));
printf('	 actual design overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UD);
// check for max. flux=Q/A=11500.(satisfactory)
Rd=((Uc-UD)/((UD)*(Uc))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.4f (hr)*(ft^2)*(F)/Btu 
',Rd);
printf('	 pressure drop  for inner pipe 
');
f=0.000168; // friction factor for reynolds number 59200, using fig.26
s=0.73;
phyt=1;
delPt=((f*(Gt^2)*(L)*(n))/(5.22*(10^10)*(D)*(s)*(phyt))); // using eq.7.45,psi
printf('	 delPt is : %.1f psi 
',delPt);
X1=0.11; // X1=((V^2)/(2*g)), for Gt 1060000,using fig.27
delPr=((4*n*X1)/(s)); // using eq.7.46,psi
printf('	 delPr is : %.1f psi 
',delPr);
delPT=delPt+delPr; // using eq.7.47,psi
printf('	 delPT is : %.1f psi 
',delPT);
printf('	 allowable delPa is negligible 
');
printf('	 pressure drop  for annulus 
');
Af=(3.14*(21.25^2-(116))/8);
printf('	 flow area : %.0f in^2 
',Af);
as=0.917; // ft^2
p=(3.14*21.25/2)+(3.14*1*116/2)+(21.25);
printf('	 wetted perimeter : %.1f in 
',p);
De=0.186; // ft
Gs=(Ws/(2*as)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.1e lb/(hr)*(ft^2) 
',Gs);
mu2=0.435; // at 315F, fig 14,lb/(ft)*(hr)
Res=((De)*(Gs)/mu2); // reynolds number
printf('	 reynolds number is : %.2e 
',Ret);
f=0.00028; // using fig.26
row=0.337; // fig 13.14
// soutlet max=0.071,
s=0.35; // using fig.6
phys=1;
delPs=((f*(Gs^2)*(L))/(5.22*(10^10)*(De)*(s)*(phys))); // using eq.7.44,psi
printf('	 delPs is : %.4f psi 
',delPs);
printf('	 allowable delPa is .25 psi 
');
//end
```

## Example 15.5: Calculation_of_a_Vertical_Thennosyphon_Reboiler.sce

```scilab
printf('	 example 15.5
');
printf('	 approximate values are mentioned in the book 
');
W=40800; // lb/hr
w=4570; // lb/hr
printf('	 1.for heat balance 
');
Ht1=241; // enthalpy of liquid at 228F, Btu/lb, fig 9
Ht2=338; // enthalpy of vapourat 228F, Btu/lb, fig 9
Q=(W*(Ht2-Ht1));
printf('	 total heat required for butane is : %.2e Btu/hr 
',Q);
l=868; // Btu/(lb), table 7
Q=((w)*(l)); // Btu/hr
printf('	 total heat required for steam is : %.2e Btu/hr 
',Q);
delt=125; // delt=LMTD, isothermal boiling, eq 5.14
// Tc and tc: Both streams are isuthermal
printf('	 trail 1 
');
A1=((Q)/((12000))); // Q/A1 =12000, first trial should always be taken for the maximum allowable flux
printf('	 A1 is : %.1e ft^2 
',A1);
a1=0.1963; // ft^2/lin ft
L=16;
N1=(A1/(L*a1)); // table 10
printf('	 number of tubes are : %.0f 
',N1);
N2=109; // assuming one tube passes, 13.25-in ID, from table 9
A2=(N2*L*a1); // ft^2
printf('	 total surface area is : %.0f ft^2 
',A2);
UD=((Q)/((A2)*(delt)));
printf('	 correct design overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UD);
// Assume 4: 1 recirculation ratio
rowv=(58/(359*(688/492)*(14.7/290))); // eq 15.18
printf('	 vapour density : %.2f lb/ft^3 
',rowv);
Vv=0.44; 
Vl=0.0372; // fig 6
W1=4*W;
printf('	 weight flow of recirculated liquid : %.3e lb/hr 
',W1);
VL=W1*Vl;
VV=W*Vv;
printf('	 volume of liquid : %.2e ft^3 
',VL);
printf('	 volume of vapour : %.3e ft^3 
',VV);
V=VL+VV;
printf('	 total volume out of reboiler : %.3e ft^3 
',V);
vo=(V/(W1+W));
printf('	 vo is : %.4f ft^3/lb 
',vo);
Pl=((2.3*16)/(144*(vo-Vl)))*(log10(vo/Vl));
printf('	 pressure leg : %.1f psi 
',Pl);
printf('	 frictional resistance 
');
Nt=109;
n=1; // number of passes
at1=0.302; // flow area,table 10, in^2
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=((W1+W)/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gt);
mu1=0.242; // at 228F, fig 14,lb/(ft)*(hr)
D=0.0517; // ft
Ret=((D)*(Gt)/mu1); // reynolds number
printf('	 reynolds number is : %.1e 
',Ret);
f=0.000127; // using fig.26
s=0.285;
phyt=1;
delPt=((f*(Gt^2)*(L)*(n))/(5.22*(10^10)*(D)*(s)*(phyt))); // using eq.7.45,psi
printf('	 delPt is : %.2f psi 
',delPt);
P=Pl+delPt;
printf('	 total resisitance : %.2f psi 
',P);
F=(16*0.43*62.5/144);
printf('	 driving force : %.2f psi 
',F);
// The resistances are greater than the hydrostatic head can provide; hence the recirculation ratio will be less than 4: 1
printf('	 trial 2 
'); // Assume 12'0' tubes and 4:1 recirculation ratio
A1=((Q)/((12000))); // Q/A1 =12000, first trial should always be taken for the maximum allowable flux
printf('	 A1 is : %.1e ft^2 
',A1);
a1=0.1963; // ft^2/lin ft
L=12;
N1=(A1/(L*a1)); // table 10
printf('	 number of tubes are : %.0f 
',N1);
N2=151; // assuming one tube passes, 15.25-in ID, from table 9
A2=(N2*L*a1); // ft^2
printf('	 total surface area is : %.0f ft^2 
',A2);
UD=((Q)/((A2)*(delt)));
printf('	 correct design overall coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',UD);
Pl=((2.3*12)/(144*(vo-Vl)))*(log10(vo/Vl));
printf('	 pressure leg : %.1f psi 
',Pl);
printf('	 frictional resistance 
');
Nt=151;
n=1; // number of passes
at1=0.302; // flow area,table 10, in^2
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=((W1+W)/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gt);
mu1=0.242; // at 228F, fig 14,lb/(ft)*(hr)
D=0.0517; // ft
Ret=((D)*(Gt)/mu1); // reynolds number
printf('	 reynolds number is : %.2e 
',Ret);
f=0.000135; // using fig.26
s=0.285;
phyt=1;
delPt=((f*(Gt^2)*(12)*(n))/(5.22*(10^10)*(D)*(s)*(phyt))); // using eq.7.45,psi
printf('	 delPt is : %.2f psi 
',delPt);
P=Pl+delPt;
printf('	 total resisitance : %.2f psi 
',P);
F=(12*0.43*62.5/144);
printf('	 driving force : %.2f psi 
',F);
// Since the driving force is slightly greater than the resistances, a recirculation ratio better than 4:1 is assured.
printf('	 hot fluid : shell side,steam 
');
ho=1500; // condensing steam
printf('	 cold fluid:inner tube side, butane 
');
jH=330; // from fig.24
Z=0.115; // Z=k*((c)*(mu1)/k)^(1/3), fig 16
Hi=((jH)*(1/D)*(Z)); //, Hi=(hi/phyt)using eq.6.15d,Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Hi);
Hio=((300)*(0.62/0.75)); //Hio=(hio/phyp), using eq.6.9
printf('	 Correct Hio to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',Hio);
Uc=((Hio)*(ho)/(Hio+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Uc);
UD=89;
Rd=((Uc-UD)/((UD)*(Uc))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.4f (hr)*(ft^2)*(F)/Btu 
',Rd);
// end
```
