## Example 7.3: Calculation_of_a_Kerosene_Crude_Oil_Exchanger.sce

```scilab
printf('	 example 7.3 
');
printf('	 approximate values are mentioned in the book 
');
T1=390; // inlet hot fluid,F
T2=200; // outlet hot fluid,F
t1=100; // inlet cold fluid,F
t2=170; // outlet cold fluid,F
W=43800; // lb/hr
w=149000; // lb/hr
printf('	 1.for heat balance 
');
printf('	 for kerosene 
');
c=0.605; // Btu/(lb)*(F)
Q1=((W)*(c)*(T1-T2)); // Btu/hr
printf('	 total heat required for kerosene is : %.1e Btu/hr 
',Q1); // calculation mistake in problem
printf('	 for crude oil 
');
c=0.49; // Btu/(lb)*(F)
Q=((w)*(c)*(t2-t1)); // Btu/hr
printf('	 total heat required for mid continent crude is : %.1e Btu/hr 
',Q); // calculation mistake in problem
delt1=T2-t1; //F
delt2=T1-t2; // F
printf('	 delt1 is : %.0f F 
',delt1);
printf('	 delt2 is : %.0f F 
',delt2);
LMTD=((delt2-delt1)/((2.3)*(log10(delt2/delt1))));
printf('	 LMTD is :%.1f F 
',LMTD);
R=((T1-T2)/(t2-t1));
printf('	 R is : %.2f 
',R);
S=((t2-t1)/(T1-t1));
printf('	 S is : %.3f 
',S);
printf('	 FT is 0.905 
'); // from fig 18
delt=(0.905*LMTD); // F
printf('	 delt is : %.0f F 
',delt);
X=((delt1)/(delt2));
printf('	 ratio of two local temperature difference is : %.3f 
',X);
Fc=0.42; // from fig.17
Kc=0.20; // crude oil controlling
Tc=((T2)+((Fc)*(T1-T2))); // caloric temperature of hot fluid,F
printf('	 caloric temperature of hot fluid is : %.0f F 
',Tc);
tc=((t1)+((Fc)*(t2-t1))); // caloric temperature of cold fluid,F
printf('	 caloric temperature of cold fluid is : %.0f F 
',tc);
printf('	 hot fluid:shell side,kerosene 
');
ID=21.25; // in
C=0.25; // clearance
B=5; // baffle spacing,in
PT=1.25;
as=((ID*C*B)/(144*PT)); // flow area,ft^2
printf('	 flow area is : %.4f ft^2 
',as);
Gs=(W/as); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gs);
mu1=0.40*2.42; // at 280F,lb/(ft)*(hr), from fig.14
De=0.99/12; // from fig.28,ft
Res=((De)*(Gs)/mu1); // reynolds number
printf('	 reynolds number is : %.2e 
',Res);
jH=93; // from fig.28
c=0.59; // Btu/(lb)*(F),at 280F,from fig.4
k=0.0765; // Btu/(hr)*(ft^2)*(F/ft), from fig.1
Pr=((c)*(mu1)/k)^(1/3); // prandelt number raised to power 1/3
printf('	 Pr is : %.3f 
',Pr);
Ho=((jH)*(k/De)*(Pr)); // H0=(h0/phya),using eq.6.15,Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Ho);
printf('	 cold fluid:inner tube side,crude oil 
');
D=0.0675; // ft
Nt=158;
n=4; // number of passes
L=16; //ft
at1=0.515; // flow area, in^2
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=(w/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gt);
mu2=3.6*2.42; // at 129F,lb/(ft)*(hr)
Ret=((D)*(Gt)/mu2); // reynolds number
printf('	 reynolds number is : %.2e 
',Ret);
jH=31; // from fig.24
c=0.49; // Btu/(lb)*(F),at 304F,from fig.4
k=0.077; // Btu/(hr)*(ft^2)*(F/ft), from fig.1
Pr=((c)*(mu2)/k)^(1/3); // prandelt number raised to power 1/3
printf('	 Pr is : %.3f 
',Pr);
Hi=((jH)*(k/D)*(Pr)*(1^0.14)); //Hi=(hi/phyp),using eq.6.15a,Btu/(hr)*(ft^2)*(F)
printf('	 Hi is : %.0f Btu/(hr)*(ft^2)*(F) 
',Hi);
ID=0.81; // ft
OD=1; //ft
Hio=((Hi)*(ID/OD)); //Hio=(hio/phyp), using eq.6.5
printf('	 Correct Hi0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',Hio);
muw=1.5*2.42; // lb/(ft)*(hr), from fig.14
phyt=(mu2/muw)^0.14;
printf('	 phyt is : %.2f 
',phyt); // from fig.24
hio=(Hio)*(phyt); // from eq.6.37
printf('	 Correct hi0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',hio);
tw=(tc)+(((Ho)/(Hio+Ho))*(Tc-tc)); // from eq.5.31
printf('	 tw is : %.0f F 
',tw);
muw=0.56*2.42; // lb/(ft)*(hr), from fig.14
phys=(mu1/muw)^0.14;
printf('	 phys is : %.2f 
',phys); // from fig.24
ho=(Ho)*(phys); // from eq.6.36
printf('	 Correct h0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',ho);
Uc=((hio)*(ho)/(hio+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',Uc);
A2=0.2618; // actual surface supplied for each tube,ft^2,from table 10
A=(Nt*L*A2); // ft^2
printf('	 total surface area is : %.0f ft^2 
',A);
UD=((Q)/((A)*(delt)));
printf('	 actual design overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UD);
Rd=((Uc-UD)/((UD)*(Uc))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.5f (hr)*(ft^2)*(F)/Btu 
',Rd);
printf('	 pressure drop  for annulus 
');
f=0.00175; // friction factor for reynolds number 25300, using fig.29
s=0.73; // for reynolds number 25300,using fig.6
Ds=21.25/12; // ft
N=(12*L/B); // number of crosses,using eq.7.43
printf('	 number of crosses are : %.0f 
',N);
delPs=((f*(Gs^2)*(Ds)*(N))/(5.22*(10^10)*(De)*(s)*(phys))); // using eq.7.44,psi
printf('	 delPs is : %.1f psi 
',delPs);
printf('	 allowable delPa is 10 psi 
');
printf('	 pressure drop  for inner pipe 
');
f=0.000285; // friction factor for reynolds number 8220, using fig.26
s=0.83;
delPt=((f*(Gt^2)*(L)*(n))/(5.22*(10^10)*(D)*(s)*(phyt))); // using eq.7.45,psi
printf('	 delPt is : %.1f psi 
',delPt);
X1=0.15; // X1=((V^2)/(2*g)), for Gt 1060000,using fig.27
delPr=((4*n*X1)/(s)); // using eq.7.46,psi
printf('	 delPr is : %.1f psi 
',delPr);
delPT=delPt+delPr; // using eq.7.47,psi
printf('	 delPT is : %.1f psi 
',delPT);
printf('	 allowable delPs is 10 psi 
');
//end
```

## Example 7.4: Calculation_of_a_Distilled_water_Raw_water_Exchanger.sce

```scilab
printf('	 example 7.4 
');
printf('	 approximate values are mentioned in the book 
');
T1=93; // inlet hot fluid,F
T2=85; // outlet hot fluid,F
t1=75; // inlet cold fluid,F
t2=80; // outlet cold fluid,F
W=175000; // lb/hr
w=280000; // lb/hr
printf('	 1.for heat balance 
');
printf('	 for distilled water 
');
c=1; // Btu/(lb)*(F)
Q=((W)*(c)*(T1-T2)); // Btu/hr
printf('	 total heat required for distilled water is : %.1e Btu/hr 
',Q);
printf('	 for raw water 
');
c=1; // Btu/(lb)*(F)
Q=((w)*(c)*(t2-t1)); // Btu/hr
printf('	 total heat required for raw water is : %.1e Btu/hr 
',Q);
delt1=T2-t1; //F
delt2=T1-t2; // F
printf('	 delt1 is : %.0f F 
',delt1);
printf('	 delt2 is : %.0f F 
',delt2);
LMTD=((delt2-delt1)/((2.3)*(log10(delt2/delt1))));
printf('	 LMTD is :%.1f F 
',LMTD);
R=((T1-T2)/(t2-t1));
printf('	 R is : %.2f 
',R);
S=((t2-t1)/(T1-t1));
printf('	 S is : %.3f 
',S);
printf('	 FT is 0.945 
'); // from fig 18
delt=(0.945*LMTD); // F
printf('	 delt is : %.2f F 
',delt);
X=((delt1)/(delt2));
printf('	 ratio of two local temperature difference is : %.3f 
',X);
Fc=0.42; // from fig.17
Kc=0.20; // crude oil controlling
Tc=((T2)+(T1))/2; // caloric temperature of hot fluid,F
printf('	 caloric temperature of hot fluid is : %.0f F 
',Tc);
tc=((t1)+(t2))/2; // caloric temperature of cold fluid,F
printf('	 caloric temperature of cold fluid is : %.1f F 
',tc);
printf('	 hot fluid:shell side,distilled water 
');
ID=15.25; // in
C=0.1875; // clearance
B=12; // baffle spacing,in
PT=0.9375;
as=((ID*C*B)/(144*PT)); // flow area,ft^2,using eq.7.1
printf('	 flow area is : %.3f ft^2 
',as);
Gs=(W/as); // mass velocity,lb/(hr)*(ft^2),using eq.7.2
printf('	 mass velocity is : %.1e lb/(hr)*(ft^2) 
',Gs);
mu1=0.81*2.42; // at 89F,lb/(ft)*(hr), from fig.14
De=0.55/12; // from fig.28,ft
Res=((De)*(Gs)/mu1); // reynolds number
printf('	 reynolds number is : %.2e 
',Res);
jH=73; // from fig.28
c=1; // Btu/(lb)*(F),at 89F,from fig.table 4
k=0.36; // Btu/(hr)*(ft^2)*(F/ft), from table 4
Pr=((c)*(mu1)/k)^(1/3); // prandelt number raised to power 1/3
printf('	 Pr is : %.3f 
',Pr);
ho=((jH)*(k/De)*(Pr)); // using eq.6.15,Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.2e Btu/(hr)*(ft^2)*(F) 
',ho);
printf('	 cold fluid:inner tube side,raw water 
');
Nt=160;
n=2; // number of passes
L=16; //ft
at1=0.334; // flow area, in^2,from table 10
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=(w/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.3e lb/(hr)*(ft^2) 
',Gt);
V=(Gt/(3600*62.5));
printf('	 V is %.1f fps 
',V);
mu2=0.92*2.42; // at 77.5F,lb/(ft)*(hr)
D=0.65/12; //ft
Ret=((D)*(Gt)/mu2); // reynolds number
printf('	 reynolds number is : %.2e 
',Ret);
hi=1350*0.99; //using fig.25,Btu/(hr)*(ft^2)*(F)
ID=0.65; // ft
OD=0.75; //ft
hio=((hi)*(ID/OD)); // using eq.6.5
printf('	 Correct hi0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',hio);
Uc=((hio)*(ho)/(hio+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Uc);
printf('	 ·when both. film coefficients are high the thermal resistance of the tube metal is not necessarily insignificant as assumed in the derivation of Eq. (6.38). For a steel 1.8 BWG tube Rm= 0.00017 and for copper Rm= 0.000017 
');
A2=0.1963; // actual surface supplied for each tube,ft^2,from table 10
A=(Nt*L*A2); // ft^2
printf('	 total surface area is : %.0f ft^2 
',A);
UD=((Q)/((A)*(delt)));
printf('	 actual design overall coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',UD);
Rd=((Uc-UD)/((UD)*(Uc))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.4f (hr)*(ft^2)*(F)/Btu 
',Rd);
printf('	 pressure drop  for annulus 
');
f=0.0019; // friction factor for reynolds number 16200, using fig.29
s=1; // for reynolds number 25300,using fig.6
Ds=15.25/12; // ft
phys=1;
N=(12*L/B); // number of crosses,using eq.7.43
printf('	 number of crosses are : %.0f 
',N);
delPs=((f*(Gs^2)*(Ds)*(N))/(5.22*(10^10)*(De)*(s)*(phys))); // using eq.7.44,psi
printf('	 delPs is : %.1f psi 
',delPs);
printf('	 allowable delPs is 10 psi 
');
printf('	 pressure drop  for inner pipe 
');
f=0.00019; // friction factor for reynolds number 36400, using fig.26
s=1;
phyt=1;
D=0.054; // ft
delPt=((f*(Gt^2)*(L)*(n))/(5.22*(10^10)*(D)*(s)*(phyt))); // using eq.7.45,psi
printf('	 delPt is : %.1f psi 
',delPt);
X1=0.33; // X1=((V^2)/(2*g)), for Gt 1060000,using fig.27
delPr=((4*n*X1)/(s)); // using eq.7.46,psi
printf('	 delPr is : %.1f psi 
',delPr);
delPT=delPt+delPr; // using eq.7.47,psi
printf('	 delPT is : %.1f psi 
',delPT);
printf('	 allowable delPT is 10 psi 
');
//end
```

```scilab
printf('	 example 7.8 
');
printf('	 approximate values are mentioned in the book 
');
T1=228; // inlet hot fluid,F
T2=228; // outlet hot fluid,F
t1=100; // inlet cold fluid,F
t2=122; // outlet cold fluid,F
W=200000; // lb/hr
w=3950; // lb/hr
printf('	 1.for heat balance 
');
printf('	 for solution 
');
c=(0.2*0.30)+(0.8*1); // bcoz of 20 percent solution,Btu/(lb)*(F)
Q1=((W)*(c)*(t2-t1)); // Btu/hr
printf('	 total heat required for solution is : %.2e Btu/hr 
',Q1);
printf('	 for steam 
');
l=960.1; // latent heat of condensation,Btu/(lb)
Q=((w)*(l)); // Btu/hr
printf('	 total heat required for steam is : %.2e Btu/hr 
',Q);
delt1=T2-t1; //F
delt2=T1-t2; // F
printf('	 delt1 is : %.0f F 
',delt1);
printf('	 delt2 is : %.0f F 
',delt2);
LMTD=((delt2-delt1)/((2.3)*(log10(delt2/delt1))));
printf('	 LMTD is :%.1f F 
',LMTD);
R=((T1-T2)/(t2-t1));
printf('	 R is : %.0f 
',R);
delt=(LMTD); // when R=0,F
printf('	 delt is : %.1f F 
',delt);
printf('	 The steam coefficient will be very great compared with that for the sugar solution, and the tube wall will be considerably nearer 228°F than the caloric temperature of the fluid. Obtain Fc from U1 and U0 Failure to correct for wall effects, however, will keep the heater calculation on the safe side.
');
ta=111; //F
Ta=228; //f
printf('	 hot fluid:tube side,steam 
');
Nt=76;
n=2; // number of passes
L=16; //ft
at1=0.302; // flow area, in^2
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.4f ft^2 
',at);
Gt=(w/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2elb/(hr)*(ft^2) 
',Gt);
mu2=0.0128*2.42; // at 228F,lb/(ft)*(hr)
D=(0.62/12); // from table 10,ft
Ret=((D)*(Gt)/mu2); // reynolds number
printf('	 reynolds number is : %.2e 
',Ret);
hio=1500; // for condensation of steam
printf('	 Correct hi0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',hio);
printf('	 cold fluid:shell side,sugar solution 
');
ID=12; // in
d=0.75/12; // diameter of tube,ft
Nt=76; // number of tubes
as=((3.14*(12^2)/4)-(76*3.14*(0.75^2)/4))/144; // flow area,ft^2
printf('	 flow area is : %.2f ft^2 
',as);
Gs=(W/as); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gs);
mu1=1.30*2.42; // at 111F,lb/(ft)*(hr), from fig.14
De=((4*as)/(Nt*3.14*d)); // from eq.6.3,ft
printf('	 De is : %.3f ft 
',De);
Res=((De)*(Gs)/mu1); // reynolds number
printf('	 reynolds number is : %.2e 
',Res);
jH=61.5; // from fig.24, tube side data
c=0.86; // Btu/(lb)*(F),at 111F,from fig.4
k=0.333; // Btu/(hr)*(ft^2)*(F/ft)
Pr=((c)*(mu1)/k)^(1/3); // prandelt number raised to power 1/3
printf('	 Pr is : %.0f 
',Pr);
Ho=((jH)*(k/De)*(Pr)); // H0=(h0/phya),using eq.6.15,Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Ho);
muw=0.51*2.42; //  at 210F,lb/(ft)*(hr), from fig.14
phys=(mu1/muw)^0.14;
printf('	 phys is : %.2f 
',phys); // from fig.24
ho=(Ho)*(phys); // from eq.6.36
printf('	 Correct h0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',ho);
tw=(ta)+(((hio)/(hio+Ho))*(Ta-ta)); // from eq.5.31
printf('	 tw is : %.0f F 
',tw);
Uc=((hio)*(ho)/(hio+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Uc);
A2=0.1963; // actual surface supplied for each tube,ft^2,from table 10
A=(Nt*L*A2); // ft^2
printf('	 total surface area is : %.0f ft^2 
',A);
UD=((Q)/((A)*(LMTD)));
printf('	 actual design overall coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',UD);
Rd=((Uc-UD)/((UD)*(Uc))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.4f (hr)*(ft^2)*(F)/Btu 
',Rd);
printf('	 pressure drop  for inner pipe 
');
f=0.000155; // friction factor for reynolds number 82500, using fig.26
s=0.0008;
phyt=1;
D=0.0517; 
delPt=((f*(Gt^2)*(L)*(2))/(5.22*(10^10)*(D)*(s)*(phyt)))/2; // using eq.7.45,psi
printf('	 delPt is : %.1f psi 
',delPt);
printf('	 pressure drop  for annulus 
');
De1=((4*as)/((Nt*3.14*d)+(3.14*1))); // from eq.6.4,ft
printf('	 De1 is : %.3f ft 
',De1);
Res1=(De1*Gs/mu1); // from eq 7.3
printf('	 Res1 is : %.2e 
',Res1);
f=0.00025; // friction factor, using fig.26
s=1.08; // for reynolds number 25300,using fig.6
delPs=((f*(Gs^2)*(L)*(1))/(5.22*(10^10)*(De1)*(s)*(phys))); // using eq.7.44,psi
printf('	 delPs is : %.2f psi 
',delPs);
//end
```
