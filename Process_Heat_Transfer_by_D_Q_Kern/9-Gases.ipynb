## Example 9.1: Calculation_of_an_Ammonia_Compressor_Aftercooler.sce

```scilab
printf('	 example 9.1 
');
printf('	 approximate values are mentioned in the book 
');
T1=245; // inlet hot fluid,F
T2=95; // outlet hot fluid,F
t1=85; // inlet cold fluid,F
t2=95; // outlet cold fluid,F
W=9872; // lb/hr
w=78500; // lb/hr
printf('	 1.for heat balance 
');
printf('	 for ammonia gas 
');
c=0.53; // Btu/(lb)*(F)
Q=((W)*(c)*(T1-T2)); // Btu/hr
printf('	 total heat required for ammonia gas is : %.2e Btu/hr 
',Q);
printf('	 for water 
');
c=1; // Btu/(lb)*(F)
Q=((w)*(c)*(t2-t1)); // Btu/hr
printf('	 total heat required for water is : %.2f Btu/hr 
',Q);
delt1=T2-t1; //F
delt2=T1-t2; // F
printf('	 delt1 is : %.0f F 
',delt1);
printf('	 delt2 is : %.0f F 
',delt2);
LMTD=((delt2-delt1)/((2.3)*(log10(delt2/delt1))));
printf('	 LMTD is :%.1f F 
',LMTD);
R=((T1-T2)/(t2-t1));
printf('	 R is : %.0f 
',R);
S=((t2-t1)/(T1-t1));
printf('	 S is : %.4f 
',S);
printf('	 FT is 0.837 
'); // from fig 18
delt=(0.837*LMTD); // F
printf('	 delt is : %.1f F 
',delt);
Tc=((T2)+(T1))/2; // caloric temperature of hot fluid,F
printf('	 caloric temperature of hot fluid is : %.0f F 
',Tc);
tc=((t1)+(t2))/2; // caloric temperature of cold fluid,F
printf('	 caloric temperature of cold fluid is : %.0f F 
',tc);
printf('	 hot fluid:shell side,ammonia at 83psia 
');
ID=23.25; // in
C=0.1875; // clearance
B=12; // baffle spacing,in
PT=0.937;
as=((ID*C*B)/(144*PT)); // flow area,ft^2,from eq 7.1
printf('	 flow area is : %.3f ft^2 
',as);
Gs=(W/as); // mass velocity,lb/(hr)*(ft^2),from eq 7.2
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gs);
mu1=0.012*2.42; // at 170F,lb/(ft)*(hr), from fig.15
De=0.55/12; // from fig.28,ft
Res=((De)*(Gs)/mu1); // reynolds number
printf('	 reynolds number is : %.2e 
',Res);
jH=118; // from fig.28
k=0.017; // Btu/(hr)*(ft^2)*(F/ft),from table 5
Z=0.97; // Z=(Pr*(1/3)) prandelt number
ho=((jH)*(k/De)*(Z)*1); // using eq.6.15,Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',ho);
printf('	 cold fluid:inner tube side,water 
');
Nt=364;
n=8; // number of passes
L=8; //ft
at1=0.302; // flow area, in^2,from table 10
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.4f ft^2 
',at);
Gt=(w/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gt);
V=(Gt/(3600*62.5)); // fps
printf('	 V is : %.2f fps 
',V);
mu2=0.82*2.42; // at 90F,lb/(ft)*(hr),from fig 14
D=(0.62/12); // ft,from table 10
Ret=((D)*(Gt)/mu2); // reynolds number
printf('	 reynolds number is : %.2e 
',Ret);
hi=900; // using fig 25,Btu/(hr)*(ft^2)*(F)
printf('	 hi is : %.0f Btu/(hr)*(ft^2)*(F) 
',hi);
ID=0.62; // ft
OD=0.75; //ft
hio=((hi)*(ID/OD)); // using eq.6.5
printf('	 Correct hi0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',hio);
Uc=((hio)*(ho)/(hio+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',Uc);
A2=0.1963; // actual surface supplied for each tube,ft^2,from table 10
A=(Nt*L*A2); // ft^2
printf('	 total surface area is : %.0f ft^2 
',A);
UD=((Q)/((A)*(delt)));
printf('	 actual design overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UD);
Rd=((Uc-UD)/((UD)*(Uc))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.3f (hr)*(ft^2)*(F)/Btu 
',Rd);
printf('	 pressure drop  for annulus 
');
f=0.00162; // friction factor for reynolds number 40200, using fig.29
Ds=23.25/12; // ft
phys=1;
N=(12*L/B); // number of crosses,using eq.7.43
printf('	 number of crosses are : %.0f 
',N);
rowgas=0.209;
printf('	 rowgas is %.3f lb/ft^3 
',rowgas);
s=rowgas/62.5;
printf('	 s is %.5f 
',s);
delPs=((f*(Gs^2)*(Ds)*(N))/(5.22*(10^10)*(De)*(s)*(phys))); // using eq.7.44,psi
printf('	 delPs is : %.0f psi 
',delPs);
printf('	 allowable delPs is 2 psi 
');
printf('	 pressure drop  for inner pipe 
');
f=0.000225; // friction factor for reynolds number 21400, using fig.26
s=1;
D=0.0517; //ft
phyt=1;
delPt=((f*(Gt^2)*(L)*(n))/(5.22*(10^10)*(D)*(s)*(phyt))); // using eq.7.45,psi
printf('	 delPt is : %.1f psi 
',delPt);
X1=0.090; // X1=((V^2)/(2*g)), for Gt 1060000,using fig.27
delPr=((4*n*X1)/(s)); // using eq.7.46,psi
printf('	 delPr is : %.1f psi 
',delPr);
delPT=delPt+delPr; // using eq.7.47,psi
printf('	 delPT is : %.1f psi 
',delPT);
printf('	 allowable delPT is 10 psi 
');
//end
```

## Example 9.2: Calculation_of_the_Heat_Load_for_an_Air_Intercooler.sce

```scilab
printf('	 example 9.2 
');
printf('	 approximate values are mentioned in the book 
');
V1=4670; // inlet air volume,cfm
Pp=0.8153; // Saturation partial pressure of water at 95F,psi,from table 7
Ps=404.3;//  Saturation specific volume of water at 95F,ft^3/lb, from table 7
printf('	 The air and water both occupy the same volume at their respective partial pressures 
');
Vw1=(V1*60/Ps); // water entering per hr,lb
printf('	 volume of water entering is : %.0f lb 
',Vw1);
printf('	 for first stage 
');
c=2.33; // compression ratio
P1=14.7; // psi
P2=(P1*c); // (c=(P2/P1)),psi
printf('	 P2 is : %.1f psi 
',P2);
gama=1.4; // for air
T1abs=95; // F
T2absr=((T1abs+460)*(P2/P1)^((gama-1)/gama));
printf('	 T2absr is : %.0f R 
',T2absr);
T2abs=(T2absr-459.67); // F
printf('	 T2abs is : %.0f F 
',T2abs);
printf('	 for intercooler 
');
V2=(V1*60*P1/P2); // ft^3/hr
printf('	 final gas volume is : %.1e ft^3/hr 
',V2);
Vw2=(V2/Ps); // water remaining in air, lb/hr
printf('	 water remaining in air is : %.0f lb/hr 
',Vw2);
C=(Vw1-Vw2); // condensation in inter cooler, lb/hr
printf('	 condensation in inter cooler is : %.0f lb/hr 
',C);
Vs=14.8; // Specific volume of atmospheric air,ft^3/lb
printf('	 Specific volume of atmospheric air is : %.1f ft^3/lb 
',Vs);
Va=(V1*60/Vs); // air in inlet gas, lb/hr
printf('	 air in inlet gas is : %.2e lb/hr
',Va);
printf('	 heat load(245 to 95F) 
)');
printf('	 sensible heat 
');
Qair=((Va)*(0.25)*(245-T1abs)); // Btu/hr
printf('	 Qair is : %.2e Btu/hr 
',Qair);
Qwaters=(Vw1*0.45*(245-T1abs)); // Btu/hr
printf('	 Qwaters is : %.2e Btu/hr 
',Qwaters);
printf('	 latent heat 
');
l=1040.1; // latent heat
Qwaterl=(C*l); // Btu/hr
printf('	 Qwater1 is : %.2e Btu/hr 
',Qwaterl);
Qt1=Qair+Qwaters+Qwaterl;
printf('	 total heat is : %.3e Btu/hr 
',Qt1);
printf('	 for second stage 
');
c=2.33; // compression ratio
P3=(P2*c); // (c=(P3/P1)),psi
printf('	 P3 is : %.1f psi 
',P3);
V3=(V1*60*P1/P3); // ft^3/hr
printf('	 final gas volume is : %.2e ft^3/hr 
',V3);
Vw3=(V3/Ps); // water remaining in air, lb/hr
printf('	 water remaining in air is : %.1f lb/hr 
',Vw3);
C1=(297-Vw3); // condensation in inter cooler, lb/hr
printf('	 condensation in inter cooler is : %.1f lb/hr 
',C1);
printf('	 heat load(245 to 95F) 
)');
printf('	 sensible heat 
');
Qair=(Va*0.25*(245-T1abs)); // Btu/hr
printf('	 Qair is : %.2e Btu/hr 
',Qair);
Qwaters=(Vw2*0.44*(245-T1abs)); // Btu/hr
printf('	 Qwater is : %.2e Btu/hr 
',Qwaters);
printf('	 latent heat 
');
l=1040.1; // latent heat
Qwaterl=(C1*l); // Btu/hr, calculation mistake in book
printf('	 Qwater is : %.2e Btu/hr 
',Qwaterl);
Qt1=Qair+Qwaters+Qwaterl;
printf('	 total heat is : %.3e Btu/hr 
',Qt1);
// end
```

## Example 9.3: Calculation_of_the_Dew_Point_after_Compression.sce

```scilab
printf('	 example 9.3 
');
printf('	 approximate values are mentioned in the book 
');
Va=18900; // air in inlet gas
Vw1=692; // water entering
Ma=(Va/29); // moles
Mw=(Vw1/18); // moles
M=(Ma+Mw); // moles
printf('	 total number of moles re : %.1f 
',M);
printf('	 Moles of air is : %.0f 
',Ma);
printf('	 Moles of water is : %.1f 
',Mw);
printf('	 after compression 
');
P=34.2; // pressure,psi
pw=(Mw/M)*(P); // partial pressure
printf('	 partial pressure is :%.1f psi 
',pw);
Td=124; // F, table table 7
printf('	 dew point is : %.0f F 
',Td);
// end
```
