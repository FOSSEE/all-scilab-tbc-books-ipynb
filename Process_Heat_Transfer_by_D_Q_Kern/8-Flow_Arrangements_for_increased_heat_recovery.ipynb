# Chapter 8: Flow Arrangements for increased heat recovery

## Example 8.1: Calculation_of_a_Oil_Cooler.sce

```scilab
printf('	 example 8.1 
');
printf('	 approximate values are mentioned in the book 
');
T1=358; // inlet hot fluid,F
T2=100; // outlet hot fluid,F
t1=90; // inlet cold fluid,F
t2=120; // outlet cold fluid,F
W=49600; // lb/hr
w=233000; // lb/hr
printf('	 1.for heat balance 
');
printf('	 for oil 
');
c=0.545; // Btu/(lb)*(F)
Q=((W)*(c)*(T1-T2)); // Btu/hr
printf('	 total heat required for oil is : %.2e Btu/hr 
',Q);
printf('	 for water 
');
c=1; // Btu/(lb)*(F)
Q=((w)*(c)*(t2-t1)); // Btu/hr
printf('	 total heat required for water is : %.2e Btu/hr 
',Q);
delt1=T2-t1; //F
delt2=T1-t2; // F
printf('	 delt1 is : %.0f F 
',delt1);
printf('	 delt2 is : %.0f F 
',delt2);
LMTD=((delt2-delt1)/((2.3)*(log10(delt2/delt1))));
printf('	 LMTD is :%.0f F 
',LMTD);
R=((T1-T2)/(t2-t1));
printf('	 R is : %.1f 
',R);
S=((t2-t1)/(T1-t1));
printf('	 S is : %.3f 
',S);
printf('	 FT is 0.93 
'); // from fig 19 for 2-4 exchanger
delt=(0.93*LMTD); // F
printf('	 delt is : %.1f F 
',delt);
X=((delt1)/(delt2));
printf('	 ratio of two local temperature difference is : %.3f 
',X);
Fc=0.25; // from fig.17
Kc=0.47; // crude oil controlling
Tc=((T2)+((Fc)*(T1-T2))); // caloric temperature of hot fluid,F
printf('	 caloric temperature of hot fluid is : %.0f F 
',Tc);
tc=((t1)+((Fc)*(t2-t1))); // caloric temperature of cold fluid,F
printf('	 caloric temperature of cold fluid is : %.0f F 
',tc);
printf('	 hot fluid:shell side,oil 
');
ID=35; // in
C=0.25; // clearance
B=7; // baffle spacing,in
PT=1.25;
as=((ID*C*B)/(144*PT))/2; // flow area,ft^2,from eq 7.1
printf('	 flow area is : %.2f ft^2 
',as);
Gs=(W/as); // mass velocity,lb/(hr)*(ft^2),from eq 7.2
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gs);
mu1=1.12*2.42; // at 165F,lb/(ft)*(hr), from fig.14
De=0.99/12; // from fig.28,ft
Res=((De)*(Gs)/mu1); // reynolds number
printf('	 reynolds number is : %.1e 
',Res);
jH=52.5; // from fig.28
Z=0.2; // Z=(k)*(Pr*(1/3)) prandelt number
Ho=((jH)*(1/De)*(Z)); // H0=(h0/phys),using eq.6.15,Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Ho);
printf('	 cold fluid:inner tube side,water 
');
Nt=454;
n=6; // number of passes
L=12; //ft
at1=0.455; // flow area, in^2
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=(w/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gt);
V=(Gt/(3600*62.5)); // fps
printf('	 V is : %.2f fps 
',V);
mu2=0.73*2.42; // at 98F,lb/(ft)*(hr),from fig 14
D=(0.76/12); // ft,from table 10
Ret=((D)*(Gt)/mu2); // reynolds number
printf('	 reynolds number is : %.2e 
',Ret);
hi=1010*0.96; // using fig 25,Btu/(hr)*(ft^2)*(F)
printf('	 hi is : %.0f Btu/(hr)*(ft^2)*(F) 
',hi);
ID=0.76; // ft
OD=1; //ft
hio=((hi)*(ID/OD)); // using eq.6.5
printf('	 Correct hi0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',hio);
tw=(tc)+(((Ho)/(hio+Ho))*(Tc-tc)); // from eq.5.31
printf('	 tw is : %.0f F 
',tw);
muw=1.95*2.42; // lb/(ft)*(hr), from fig.14
phys=(mu1/muw)^0.14;
printf('	 phys is : %.2f 
',phys); // from fig.24
ho=(Ho)*(phys); // from eq.6.36
printf('	 Correct h0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',ho);
Uc=((hio)*(ho)/(hio+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',Uc);
A2=0.2618; // actual surface supplied for each tube,ft^2,from table 10
A=(Nt*L*A2); // ft^2
printf('	 total surface area is : %.0f ft^2 
',A);
Q=6980000; // taking rounded value,Btu/hr
UD=((Q)/((A)*(delt)));
printf('	 actual design overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UD);
Rd=((Uc-UD)/((UD)*(Uc))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.4f (hr)*(ft^2)*(F)/Btu 
',Rd);
printf('	 pressure drop  for annulus 
');
f=0.00215; // friction factor for reynolds number 8900, using fig.29
s=0.82; // for reynolds number 25300,using fig.6
Ds=35/12; // ft
N=(12*L/B); // number of crosses,using eq.7.43
printf('	 number of crosses are : %.0f 
',N);
delPs=((f*(Gs^2)*(Ds)*(2*N))/(5.22*(10^10)*(De)*(s)*(phys))); // using eq.7.44,psi
printf('	 delPs is : %.0f psi 
',delPs);
printf('	 allowable delPs is 10 psi 
');
printf('	 pressure drop  for inner pipe 
');
f=0.000195; // friction factor for reynolds number 34900, using fig.26
s=1;
D=0.0633; //ft
phyt=1;
delPt=((f*(Gt^2)*(L)*(n))/(5.22*(10^10)*(D)*(s)*(phyt))); // using eq.7.45,psi
printf('	 delPt is : %.1f psi 
',delPt);
X1=0.13; // X1=((V^2)/(2*g)), for Gt 1060000,using fig.27
delPr=((4*n*X1)/(s)); // using eq.7.46,psi
printf('	 delPr is : %.1f psi 
',delPr);
delPT=delPt+delPr; // using eq.7.47,psi
printf('	 delPT is : %.1f psi 
',delPT);
printf('	 allowable delPT is 10 psi 
');
//end
```

## Example 8.2: calculation_of_an_Acetone_Acetic_Acid_Exchanger.sce

```scilab
printf('	 example 8.2 
');
printf('	 approximate values are mentioned in the book 
');
T1=250; // inlet hot fluid,F
T2=100; // outlet hot fluid,F
t1=90; // inlet cold fluid,F
t2=150; // outlet cold fluid,F
W=60000; // lb/hr
w=168000; // lb/hr
printf('	 1.for heat balance 
');
printf('	 for acetone 
');
c=0.57; // Btu/(lb)*(F)
Q=((W)*(c)*(T1-T2)); // Btu/hr
printf('	 total heat required for acetone is : %.2e Btu/hr 
',Q); // calculation mistake in problem
printf('	 for acetic acid 
');
c=0.51; // Btu/(lb)*(F)
Q1=((w)*(c)*(t2-t1)); // Btu/hr
printf('	 total heat required for acetic acid is : %.2e Btu/hr 
',Q1); // calculation mistake in problem
delt1=T2-t1; //F
delt2=T1-t2; // F
printf('	 delt1 is : %.0f F 
',delt1);
printf('	 delt2 is : %.0f F 
',delt2);
LMTD=((delt2-delt1)/((2.3)*(log10(delt2/delt1))));
printf('	 LMTD is :%.1f F 
',LMTD);
R=((T1-T2)/(t2-t1));
printf('	 R is : %.1f 
',R);
S=((t2-t1)/(T1-t1));
printf('	 S is : %.3f 
',S);
printf('	 FT is 0.88 
'); // from fig 20,for 3-6 exchanger
delt=(0.88*LMTD); // F
printf('	 delt is : %.1f F 
',delt);
Tc=((T2)+(T1))/2; // caloric temperature of hot fluid,F
printf('	 caloric temperature of hot fluid is : %.0f F 
',Tc);
tc=((t1)+(t2))/2; // caloric temperature of cold fluid,F
printf('	 caloric temperature of cold fluid is : %.0f F 
',tc);
printf('	 hot fluid:shell side,acetone 
');
ID=21.25; // in
C=0.25; // clearance
B=5; // baffle spacing,in
PT=1;
as=((ID*C*B)/(144*PT)); // flow area,ft^2
printf('	 flow area is : %.3f ft^2 
',as);
Gs=(W/as); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.2e lb/(hr)*(ft^2) 
',Gs);
mu1=0.20*2.42; // at 175F,lb/(ft)*(hr), from fig.14
De=0.95/12; // from fig.28,ft
Res=((De)*(Gs)/mu1); // reynolds number
printf('	 reynolds number is : %.2e 
',Res);
phys=1;
jH=137; // from fig.28
c=0.63; // Btu/(lb)*(F),at 175F,from fig.2
k=0.095; // Btu/(hr)*(ft^2)*(F/ft), from table 4
Pr=((c)*(mu1)/k)^(1/3); // prandelt number raised to power 1/3
printf('	 Pr is : %.3f 
',Pr);
ho=((jH)*(k/De)*(Pr)*1); // using eq.6.15b,Btu/(hr)*(ft^2)*(F)
printf('	 individual heat transfer coefficient is : %.0f Btu/(hr)*(ft^2)*(F) 
',ho);
printf('	 cold fluid:inner tube side,acetic acid 
');
Nt=270;
n=2; // number of passes
L=16; //ft
at1=0.268; // flow area, in^2,from table 10
at=((Nt*at1)/(144*n)); // total area,ft^2,from eq.7.48
printf('	 flow area is : %.3f ft^2 
',at);
Gt=(w/(at)); // mass velocity,lb/(hr)*(ft^2)
printf('	 mass velocity is : %.1e lb/(hr)*(ft^2) 
',Gt);
mu2=0.85*2.42; // at 120F,lb/(ft)*(hr)
D=(0.584/12); // ft
Ret=((D)*(Gt)/mu2); // reynolds number
printf('	 reynolds number is : %.2e 
',Ret);
jH=56; // from fig.24
c=0.51; // Btu/(lb)*(F),at 120F,from fig.2
k=0.098; // Btu/(hr)*(ft^2)*(F/ft), from table 4
phyt=1;
Pr=((c)*(mu2)/k)^(1/3); // prandelt number raised to power 1/3
printf('	 Pr is : %.3f 
',Pr);
hi=((jH)*(k/D)*(Pr)*(1^0.14)); // using eq.6.15a,Btu/(hr)*(ft^2)*(F)
printf('	 hi is : %.0f Btu/(hr)*(ft^2)*(F) 
',hi);
ID=0.584; // ft
OD=0.75; //ft
hio=((hi)*(ID/OD)); //Hio=(hio/phyp), using eq.6.5
printf('	 Correct hi0 to the surface at the OD is : %.0f Btu/(hr)*(ft^2)*(F) 
',hio);
Uc=((hio)*(ho)/(hio+ho)); // clean overall coefficient,Btu/(hr)*(ft^2)*(F)
printf('	 clean overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',Uc);
A2=0.1963; // actual surface supplied for each tube,ft^2,from table 10
A=3*(Nt*L*A2); // ft^2
printf('	 total surface area is : %.2e ft^2 
',A);
UD=((Q)/((A)*(delt)));
printf('	 actual design overall coefficient is : %.1f Btu/(hr)*(ft^2)*(F) 
',UD);
Rd=((Uc-UD)/((UD)*(Uc))); // (hr)*(ft^2)*(F)/Btu
printf('	 actual Rd is : %.4f (hr)*(ft^2)*(F)/Btu 
',Rd);
printf('	 pressure drop  for annulus 
');
f=0.00155; // friction factor for reynolds number 52200, using fig.29
s=0.79; // for reynolds number 25300,using table.6
Ds=21.25/12; // ft
N=(12*L/B)+1; // number of crosses,using eq.7.43
printf('	 number of crosses are : %.0f 
',N);
delPs=((f*(Gs^2)*(Ds)*(3*N))/(5.22*(10^10)*(De)*(s)*(phys))); // using eq.7.44,psi
printf('	 delPs is : %.1f psi 
',delPs);
printf('	 allowable delPs is 10 psi 
');
printf('	 pressure drop  for inner pipe 
');
f=0.00024; // friction factor for reynolds number 158000, using fig.26
s=1.07;
D=0.0487; // ft
delPt=((f*(Gt^2)*(L)*(n*3))/(5.22*(10^10)*(D)*(s)*(phyt))); // using eq.7.45,psi
printf('	 delPt is : %.1f psi 
',delPt);
X1=0.063; // X1=((V^2)/(2*g)), for Gt 1060000,using fig.27
delPr=(3)*((4*n*X1)/(s)); // using eq.7.46,psi
printf('	 delPr is : %.1f psi 
',delPr);
delPT=delPt+delPr; // using eq.7.47,psi
printf('	 delPT is : %.1f psi 
',delPT);
printf('	 allowable delPT is 10 psi 
');
//end
```
