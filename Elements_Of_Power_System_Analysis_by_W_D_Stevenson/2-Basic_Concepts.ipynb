## Example 1: problem.on_power.sce

```scilab
//Chapter 2
//Example 2.1 
//Page 21
clear;clc;
E1=100+0*%i;
E2=86.6+50*%i;
Z=5*%i;
X=5;
printf('

 Value of voltage source one designated as a machine 1 =  %.4f V 

',E1)
printf('

 Value of voltage source two designated as a machine 2 =  %.4f V 

',E2)
printf('

 Impedance connected  = %.4f ohms 

',abs(Z))
//Calculation Of Current
I=(E1-E2)/Z;
printf('

 Current through the impedance = %.4f A 

',I)
//Calculation Of Power
M1=E1*conj(I);
disp(M1,'Machine One Power = ')
M2=E2*conj(I);
disp(M2,'Machine Two Power = ')
//Calculation Of Reactive Power Required By Inductive Reactance
RP=(abs(I))^2*X;
printf('

 Reactive power required by inductive reactance i.e, impedance = %.4f VAR 

',RP)
printf('

 Machine 1 consumes energy at the rate of %.4f W 

',abs(real(M1)))
printf('

 Machine 2 generates energy at the rate of %.4f W 

',abs(real(M2)))
printf('

 Machine 1 supplies reactive power at the rate of  %.4f VAR 

',imag(M1))
printf('

 Machine 2 supplies reactive power at the rate of %.4f VAR 

',abs(imag(M2)))
printf('

 Reactive power required by inductive reactance i.e, impedance = Sum of reactive power supplied by machine 1 + reactive power supplied by machine 2  = %.4f VAR 

',RP)
disp('Real Power consumed by impedance is Zero')
disp('The real power generated by machine two is transferred to machine one')
```

## Example 2.2: PhaseProb.sce

```scilab
//Chapter 2
//Example 2.2
//PhaseProb
//Page 26
clear;clc;
//Given line-line voltages
Vab = 173.2 * (cos (0)+ sin(0) * %i);
Vbc = 173.2 * (cos (240*%pi/180)+ sin(240*%pi/180) * %i);
Vca = 173.2 * (cos (120*%pi/180)+ sin(120*%pi/180) * %i);
disp('The given line-line voltages are')
//'/_' this symbol has been used to show angle
printf('

 Vab = %.4f  /_ %.4f V 

',abs(Vab),((atan(imag(Vab),real(Vab)))*180/%pi))
printf('

 Vbc = %.4f  /_ %.4f V 

',abs(Vbc),((atan(imag(Vbc),real(Vbc)))*180/%pi))
printf('

 Vca = %.4f  /_ %.4f V 

',abs(Vca),((atan(imag(Vca),real(Vca)))*180/%pi))
//Calculation of line-neutral voltage
Van = (Vab / sqrt(3)/(0.866+0.5*%i));
Vbn = (Vbc / sqrt(3)/(0.866+0.5*%i));
Vcn = (Vca / sqrt(3)/(0.866+0.5*%i));
disp('The line-neutral voltages are')
printf('

 Van = %.4f  /_ %.4f V 

',abs(Van),((atan(imag(Van),real(Van)))*180/%pi))
printf('

 Vbn = %.4f  /_ %.4f V 

',abs(Vbn),((atan(imag(Vbn),real(Vbn)))*180/%pi))
printf('

 Vcn = %.4f  /_ %.4f V 

',abs(Vcn),((atan(imag(Vcn),real(Vcn)))*180/%pi))
ZL = 10 * (cos(20*%pi/180) + sin(20*%pi/180) * %i);
printf('

 Load Impedance ZL = %.4f  /_ %.4f ohms 

',abs(ZL),((atan(imag(ZL),real(ZL)))*180/%pi))
//Calculation of line-neutral current
Ian = Van / ZL;
Ibn = Vbn / ZL;
Icn = Vcn / ZL;
disp('The resulting current in each phase')
printf('

 Ian = %.4f  /_ %.4f A 

',abs(Ian),((atan(imag(Ian),real(Ian)))*180/%pi))
printf('

 Ibn = %.4f  /_ %.4f A 

',abs(Ibn),((atan(imag(Ibn),real(Ibn)))*180/%pi))
printf('

 Icn = %.4f  /_ %.4f A 

',abs(Icn),((atan(imag(Icn),real(Icn)))*180/%pi))
```

## Example 2.3: SubstationBus.sce

```scilab
//Chapter 2
//Example 2.3 
//SubstationBus
//Page 28
clear;clc;
//Given values
Vll = 4.4e3;
Vln = Vll / sqrt(3);
Zline = 1.4 * (cos(75 * %pi / 180) + %i * sin (75 * %pi / 180));
Van = Vln * (cos(0) + %i * sin(0));
Zload = 20 * (cos(30 * %pi / 180) + %i * sin(30 * %pi / 180));
printf('

 Given line-line voltage = %.4f V 

',Vll)
printf('

 Line-neutral voltage = %.4f V 

',Vln)
//'/_' this symbol has been used to specify angle
printf('

 Impedance of the load = %.4f  /_ %.4f ohms 

',abs(Zload),((atan(imag(Zload),real(Zload)))*180/%pi))
printf('

 Impedance of the line = %.4f  /_ %.4f ohms 

',abs(Zline),((atan(imag(Zline),real(Zline)))*180/%pi))
//Calculation of phase current
Ian = Van / Zload;
printf('

 Van = %.4f  /_ %.4f V 

',abs(Van),((atan(imag(Van),real(Van)))*180/%pi))
printf('

 Ian = %.4f  /_ %.4f V 

',abs(Ian),((atan(imag(Ian),real(Ian)))*180/%pi))
//Calculation of line-neutral voltage at the substation
Vltn = Van + Ian * Zline;
//Magnitude of the voltage at the substation bus
magVl = sqrt(3) * abs(Vltn);
printf('

 The line to neutral voltage at the substation = %.4f /_ %.4f V 

',abs(Vltn),((atan(imag(Vltn),real(Vltn)))*180/%pi))
printf('

 The magnitude of the voltage at the substation bus = %.4f V',magVl)
```

## Example 2.4: Ex3inPerUnit.sce

```scilab
//Chapter 2
//Example 2.4 
//Ex3inPerUnit
//Page 32
clear;clc;
//Given values
Vll = 4.4e3;
Zline = 1.4 * (cos(75 * %pi / 180) + %i * sin (75 * %pi / 180));
Zload = 20 * (cos(30 * %pi / 180) + %i * sin(30 * %pi / 180));
Vbase = Vll;
Ibase = 127;
Zbase = (Vbase / sqrt(3)) / Ibase ;
//Displaying the given values and the base values
printf('

 Given line-line voltage = %.4f V 

',Vll)
printf('

 Impedance of the load = %.4f  /_ %.4f ohms 

',abs(Zload),((atan(imag(Zload),real(Zload)))*180/%pi))
printf('

 Impedance of the line = %.4f  /_ %.4f ohms 

',abs(Zline),((atan(imag(Zline),real(Zline)))*180/%pi))
printf('

 Base Voltage = %.4f V 

 ',Vbase)
printf('

 Base Current = %.4f V 

',Ibase)
printf('

 Base Impedance = %.4f V 

',Zbase)
Van = (Vll / sqrt(3)) * (cos(0) + %i * sin(0));
Ian = Van / Zload;
printf('

 Van = %.4f  /_ %.4f V 

',abs(Van),((atan(imag(Van),real(Van)))*180/%pi))
printf('

 Ian = %.4f  /_ %.4f V 

',abs(Ian),((atan(imag(Ian),real(Ian)))*180/%pi))
//Calculation of per-unit quantities
V_pu = Vll / Vbase;
I_pu = Ian / Ibase;
Zline_pu = Zline / Zbase;
Zload_pu = Zload / Zbase;
Van_pu = V_pu + I_pu * Zline_pu;
disp('Per-unit Quantities')
printf('

 Per Unit line-line voltage = %.4f  /_ %.4f per unit 

',abs(V_pu),((atan(imag(V_pu),real(V_pu)))*180/%pi))
printf('

 Per Unit line-neutral current = %.4f  /_ %.4f per unit 

',abs(I_pu),((atan(imag(I_pu),real(I_pu)))*180/%pi))
printf('

 Per Unit line-neutral voltage = %.4f  /_ %.4f per unit 

',abs(Van_pu),((atan(imag(Van_pu),real(Van_pu)))*180/%pi))
printf('

 Per Unit line impedance  = %.4f  /_ %.4f per unit 

',abs(Zline_pu),((atan(imag(Zline_pu),real(Zline_pu)))*180/%pi))
printf('

 Per Unit load impedance = %.4f  /_ %.4f per unit 

',abs(Zload_pu),((atan(imag(Zload_pu),real(Zload_pu)))*180/%pi))
//Calculation of line-neutral and line-line voltage
Vln = abs(Van_pu) * Vll / sqrt(3);
VLL = abs(Van_pu) * Vll;
printf('

 The line to neutral voltage at the substation , VLN = %.4f /_ %.4f V 

',abs(Vln),((atan(imag(Vln),real(Vln)))*180/%pi))
printf('

 The magnitude of the voltage at the substation bus , VLL= %.4f V',VLL)
```

## Example 2.5: PerUnit.sce

```scilab
//Chapter 2
//Example 2.5 
//PerUnit
//Page 34
clear;clc;
pu_Z_given = 0.25;
base_kV_given = 18;
base_kV_new = 20;
base_kVA_new = 100;
base_kVA_given = 500;
X11=(pu_Z_given * (base_kV_given/base_kV_new)^2 * (base_kVA_new/base_kVA_given));
disp('Per-Unit Znew = per-unit Zgiven * (base kVgiven/base kVnew)^2 * (base kVAnew/base kVAgiven)')
disp(' = 0.25 * (18/20)^2 * (100/500)')
printf('
 Per-Unit Znew = %f per unit',X11)
```
