## Example 5.8: cmp_command.sce

```scilab
clear
flag=1
clc
mode(-1)

printf('Example 8    :               Show the method of comparing files using cmp command 
')
disp('****************************************************************')
disp('Answer    :   ')
disp('INSTRUCTIONS   : ')
printf('
Here all instructions are preloaded in the form of a demo
PRESS ENTER AFTER EACH COMMAND to see its RESULT
PRESS ENTER AFTER EACH RESULT TO GO TO THE NEXT COMMAND
')
halt('.............Press [ENTER] to continue.....')
halt('')
clc
printf('	UNIX SHELL SIMULATOR(DEMO VERSION WITH PRELOADED COMMANDS)


')


printf('
# Enter the name of the two files which you want to compare 
')
fil(1)=input('','s')
fil(2)=input('','s')
printf('

 $ cmp   %s  %s     #to compare the files %s and %s
',fil(1),fil(2),fil(1),fil(2))

for i=1:2
if ~isfile(fil(i)) then
    flag(i)=0
    printf('
%s : file not found 
',fil(i))
    printf('# Create a new file named %s?
 # y : Yes 
 # n : No 
',fil(i))
    resp=input('','s')
    if resp=='y' then
        flag(i)=1
        printf('
#***Enter the contents of the file %s*****
# [Enter ^ in a newline to end and close the file]
',fil(i))
        printf('

$ cat > %s                      #to create a file named %s and fill its contents
',fil(i),fil(i))
        fhdr=mopen(fil(i),'wt')
        count=1
        while %t
            cont=input(string(count)+'.  ','s')
            if (cont=='^') then
                break
            end
            mfprintf(fhdr,'%s
',cont)
            count=count+1
        end
        mclose(fhdr)
    end
end
end


if flag(1)&flag(2) then
    clc
    printf('
 $ cmp   %s  %s    #to compare files %s and %s 
',fil(1),fil(2),fil(1),fil(2))
    fhdr1=mopen(fil(1),'rt')
    fhdr2=mopen(fil(2),'rt')
    l=0
    cr=1
    while %t
        [n,a1]=mfscanf(fhdr1,'%c')
        [n,a2]=mfscanf(fhdr2,'%c')
        if meof(fhdr1)&meof(fhdr2) then
           printf('  
# No output means both the files are identical 
')
            break
         elseif (meof(fhdr1)&~meof(fhdr2))|(meof(fhdr2)&~meof(fhdr1))|a1~=a2
          printf(' %s  %s  differ :  char %d ,  line %d 
',fil(1),fil(2),cr,l+1)
          printf(' # This shows that %dth character in %dth line do not match

',cr,l+1)
          break
         end
        cr=cr+1
        if ascii(a1)==10 then
            l=l+1
            cr=1
        end
    end
    mclose(fhdr1)
    mclose(fhdr2)
    halt('')
else
    printf('

# file %s or %s is not found 
',fil(1),fil(2))
end


printf('


$ exit        #To exit the current simulation terminal and return to Scilab console

')
halt('........# (hit [ENTER] for result)')
//clc()

printf('

			BACK TO SCILAB CONSOLE...
Loading initial environment')
sleep(1000)
```
