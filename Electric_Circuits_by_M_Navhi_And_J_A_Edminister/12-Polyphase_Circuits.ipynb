# Chapter 12: Polyphase Circuits

## Example 12.2: Balanced_Delta_connected_load.sce

```scilab
clc
disp('Example 12.2')
printf('
')

printf('Given')
disp('The system ABC is DELTA connected')
disp('Effective line voltage is 120V')
disp('The three impedances are 5(45 deg)')
Zmag=5;Zph=45;
//Let maximum line voltage is Vmax
Vmax=120*sqrt(2)
//From fig 12.7(a)
//VAB=Vmax(120 deg)
//VBC=Vmax(0 deg)
//VCA=Vmax(240 deg)

//From figure 12.8
IABmag=Vmax/Zmag
IABph=120-Zph
printf('IAB=%3.2f(%d deg)
',IABmag,IABph);

IBCmag=Vmax/Zmag
IBCph=0-Zph
printf('IBC=%3.2f(%d deg)
',IBCmag,IBCph);

ICAmag=Vmax/Zmag
ICAph=240-Zph
printf('ICA=%3.2f(%d deg)
',ICAmag,ICAph);

//Applying KCL equation 
//IA=IAB+IAC
//IB=IBC+IBA
//IC=ICA+ICB

x=IABmag*cos((IABph*%pi)/180);
y=IABmag*sin((IABph*%pi)/180);
z=complex(x,y)

x1=ICAmag*cos((ICAph*%pi)/180);
y1=ICAmag*sin((ICAph*%pi)/180);
z1=complex(x1,y1)

x2=IBCmag*cos((IBCph*%pi)/180);
y2=IBCmag*sin((IBCph*%pi)/180);
z2=complex(x2,y2)

IA=z-z1;
[RA,ThetaA]=polar(IA)

IB=z2-z;
[RB,ThetaB]=polar(IB)

IC=z1-z2
[RC,ThetaC]=polar(IC)

disp('Therefore')

printf('
IA=%3.2f(%d deg)A
',RA,ThetaA*(180/%pi));
printf('
IB=%3.2f(%d deg)A
',RB,ThetaB*(180/%pi));
printf('
IC=%3.2f(%d deg)A
',RC,ThetaC*(180/%pi));
```

## Example 12.6: Unbalanced_Wye_connected_load.sce

```scilab
clc
disp('Example 12.6')
printf('
')

printf('Given')
disp('The system CBA is WYE connected')
disp('Maximum line voltage is 150V')
disp('The three impedances are 6(0 deg),6(30 deg),5(45 deg)')
ZAmag=6;ZAph=0;
ZBmag=6;ZBph=30;
ZCmag=5;ZCph=45;
//Let maximum line voltage is Vmax
Vmax=150
//Let the line to neutral voltage magnitude be Vn
Vn=Vmax/sqrt(3)
//From fig 12.7(b)
//VAN=Vn(-90 deg)
//VBN=Vn(30 deg)
//VCN=Vn(150 deg)

//From figure 12.16
IAmag=Vn/ZAmag
IAph=-90-ZAph
printf('
IA=%3.2f(%d deg)A
',IAmag,IAph);

IBmag=Vn/ZBmag
IBph=30-ZBph
printf('
IB=%3.2f(%d deg)A
',IBmag,IBph);

ICmag=Vn/ZCmag
ICph=150-ZCph
printf('
IC=%3.2f(%d deg)A
',ICmag,ICph);

//Now to calculate IN
//IN=-(IA+IB+IC)
x=IAmag*cos((IAph*%pi)/180);
y=IAmag*sin((IAph*%pi)/180);
z=complex(x,y)

x1=ICmag*cos((ICph*%pi)/180);
y1=ICmag*sin((ICph*%pi)/180);
z1=complex(x1,y1)

x2=IBmag*cos((IBph*%pi)/180);
y2=IBmag*sin((IBph*%pi)/180);
z2=complex(x2,y2)

IN=-(z+z1+z2)

[R,Theta]=polar(IN)

printf('
IN=%3.2f(%d deg)A
',R,Theta*(180/%pi));
```
