## Example 11.10: LU_Method.sce

```scilab
//Example 11.10
//LU Method
//Page no. 363
clc;close;clear;

A=[120,80,40,-16;80,120,16,-40;40,16,120,-80;-16,-40,-80,120];
disp(A,'A =')
L=eye(4,4);
for l=1:20
for j=1:4
    for i=1:j
        k=0
        for p=1:i-1
            k=k-A(i,p)*A(p,j)
        end
        A(i,j)=A(i,j)+k
    end
    for i=j+1:4
        k=0;
        for p=1:j-1
            k=k-A(i,p)*A(p,j)
        end
        A(i,j)=(A(i,j)+k)/A(j,j) 
    end
end
disp(A,'Modified A = ')
    for i=1:4
    for j=1:4
        if i>j then
            L(i,j)=A(i,j)
        else
            U(i,j)=A(i,j)
        end
    end
end
disp(U,'U =',L,'L =')
A=U*L;
printf('

After %i iterations, matrix A =

',l)
for i=1:4
    for j=1:4
        printf('    %.2f	',A(i,j))
    end
    printf('
')
end
end
printf('

Therefore the eigenvalues are the diagonal elements f the transformed triangular matrix are:

')
for i=1:4
    printf(' %.2f,',A(i,i))
end
```

## Example 11.11: Generalized_Eigenvalue_Problem.sce

```scilab
//Example 11.11
//Generalized Eigenvalue Problem
//Page no. 365
clc;close;clear;

A=[1,1,0.5;1,1,0.25;0.5,0.25,2]
B=[2,2,2;2,5,5;2,5,11]
disp(B,'B =',A,'A =')
for i=1:3
    G(i,i)=sqrt(B(i,i))
end
G=[B;eye(3,3)];

//transformation to frobenius matrix
for k=3:-1:2
    g(k)=0;
    for j=1:k-1
        if(g(k)<G(k,j))
            g(k)=G(k,j)
            p=j;
        end
    end
    if(g(k)~=0)
        for j=1:3
            r(1,j)=G(k,j)
        end
        for i=1:6
            G(i,k-1)=G(i,k-1)/g(k)
        end
        for j=1:3
            if(j~=k-1)
                l=G(k,j)
                for i=1:6
                    G(i,j)=G(i,j)-l*G(i,k-1)
                end
            end
        end
    end
    for j=1:3
        for i=1:3
            c(i,1)=G(i,j)
        end
        G(k-1,j)=0
        for i=1:3
            G(k-1,j)=G(k-1,j)+r(1,i)*c(i,1)
        end
    end
end

//partition g
for i=4:6
    for j=1:3
        T(i-3,j)=G(i,j)
    end
end

//eigenvalues computation
p=poly(B,'x')
a=roots(p)
printf('

Diagonalized Matrix B = 

')
for i=1:3
    for j=1:3
        if i~=j then
            B(i,j)=0
        else
            B(i,j)=a(i)
        end
    end
end
disp(B)
//eigenvectors computation
for k=1:3
    m=2
    for l=1:3
        y(l,k)=a(k)^(m)
        m=m-1;
    end
end
printf('

')


for k=1:3
    for l=1:3
        y1(l,1)=y(l,1)
        y2(l,1)=y(l,2)
        y3(l,1)=y(l,3)
    end
    x1=T*y3;
    x2=T*y2;
    x3=T*y1;
end
printf('

Eigenvectors of B are :

')
for i=1:3
    printf('|%.5f|		|%.5f|		|%.5f|',x3(i,1),x2(i,1),x1(i,1))
    printf('
')
end
x=[x3,x2,x1]





B=[2,2,2;2,5,5;2,5,11]
G=0
for i=1:3
    for j=1:3
        if i==j then
            G(i,j)=sqrt(B(i,j))
        else
            G(i,j)=0;
        end
    end
end

B=inv(G)*x'*A*x*inv(G)
disp(B,'Eigenvectors of A =')

printf('


Note : Computation Error in book in caculation of eigenvector of B thus for A')
```

## Example 11.3: Danilevsky_Method.sce

```scilab
//Example 11.3
//Danilevsky Method
//Page no. 341
clc;close;clear;

A=[-1,0,0;1,-2,3;0,2,-3];
G=[A;eye(3,3)];
disp(G);
//transformation to frobenius matrix
for k=3:-1:2
    g(k)=0;
    for j=1:k-1
        if(g(k)<G(k,j))
            g(k)=G(k,j)
            p=j;
        end
    end
    if(g(k)~=0)
        for j=1:3
            r(1,j)=G(k,j)
        end
        for i=1:6
            G(i,k-1)=G(i,k-1)/g(k)
        end
        disp(G)
        for j=1:3
            if(j~=k-1)
                l=G(k,j)
                for i=1:6
                    G(i,j)=G(i,j)-l*G(i,k-1)
                end
            end
        end
        disp(G)
    end
    for j=1:3
        for i=1:3
            c(i,1)=G(i,j)
        end
        G(k-1,j)=0
        for i=1:3
            G(k-1,j)=G(k-1,j)+r(1,i)*c(i,1)
        end
    end
    disp(G)
end

//partition g
for i=4:6
    for j=1:3
        T(i-3,j)=G(i,j)
    end
end
disp(T,'T=')

//eigenvalues computation
printf('


Charateristic polynomial:')
p=poly(A,'x')
disp(p)
printf('


Eigenvalues:')
a=roots(p)
disp(a')
//eigenvectors computation
for k=1:3
    m=2
    for l=1:3
        y(l,k)=a(k,1)^(m)
        m=m-1;
    end
end
printf('

')
disp(y,'y=')

//eigenvector computation

for k=1:3
    for l=1:3
        y1(l,1)=y(l,1)
        y2(l,1)=y(l,2)
        y3(l,1)=y(l,3)
    end
    x1=T*y3;
    x2=T*y2;
    x3=T*y1;
end
printf('

Eigenvectors :
')
for i=1:3
    printf('|%.1f|		|%.1f|		|%.1f|',x1(i,1),x2(i,1),x3(i,1))
    printf('
')
end
```

## Example 11.7: Jacobi_Method.sce

```scilab
//Example 11.7
//Jacobi's Method
//Page no. 355
clc;close;clear;

A=[1,1,1/2;1,1,1/4;1/2,1/4,2];
C=A;
V=[sqrt(2),0,1/2;sqrt(2),0,1/4;3/(4*sqrt(2)),-1/(4*sqrt(2)),2]
S=eye(3,3)
disp(A,'A =')
VI=0;
for i=1:3
    for j=1:3
        if(i~=j)
            VI=VI+A(i,j)^2                          //initial off diag norm
        end
    end
end
VI=sqrt(VI);
VF=VI*10^-7;       //final threshold
V1=VI/3;
o=poly(0,'o');
for i=1:3
for q=2:3
    for p=q-1:-1:1
        if(A(p,q)>V1)
            a=-A(p,q);
            b=(A(p,p)-A(q,q))/2
            if(b~=0)
                w=b*abs(1/b)*(a/sqrt(a^2+b^2));
            else
                w=(a/sqrt(a^2+b^2));
            end
            sin0=w/sqrt(2*(1+sqrt(1-w^2)));
            cos0=sqrt(1-sin0^2)
        end
        B(p,p)=A(p,p)*cos0^2+A(q,q)*sin0^2-2*A(p,q)*sin0*cos0
            B(q,q)=A(p,p)*sin0^2+A(q,q)*cos0^2+2*A(p,q)*sin0*cos0
            B(p,q)=(A(p,p)-A(q,q))*sin0*cos0+A(p,q)*(cos0^2-sin0^2)
            S(i,i)=S(i,i)
            S(i,p)=S(i,p)*cos0-S(i,q)*sin0
            S(i,q)=S(i,p)*sin0+S(i,q)*cos0
            
    end
end
end
disp(B,'B =')
disp(S,'S =')
printf('


Computation error in the solution provided by book')
```
