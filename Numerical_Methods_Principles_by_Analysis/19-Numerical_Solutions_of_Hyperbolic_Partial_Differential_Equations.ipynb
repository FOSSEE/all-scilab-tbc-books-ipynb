## Example 19.10: Non_Linear_1st_Order_Hyperboolic_Differential_Equation.sce

```scilab
//Example 19.10
//Non Linear 1st Order Hyperboolic Differential Equation
//Page no. 667
clc;clear;close;

c=-2;k=0.05;h=0.2;
r=abs(c)*k/h;
printf('
 i	 x	|	j -->	')
for i=0:6
    printf('  %i	',i)
end
printf('
 |	 |	|	t -->	')
for i=0:6
    printf('%.3f	',i*k)
end
i=1;
x=0;
printf('
---------------------------------------------------------------------------------------')
for j=1:7
    for i=1:6
        if j==1 then
            u(i,j)=exp(-x);
            u(i+1,j)=exp(-(x+h));
        elseif i==1 then
            u(i,j)=1
        else
            u(i,j)=u(i,j-1)-k*(u(i+1,j-1)^2-u(i-1,j-1)^2)/(4*h)+k^2*((u(i+1,j-1)+u(i,j-1))*(u(i+1,j-1)^2-u(i,j-1)^2)-(u(i,j-1)+u(i-1,j-1))*(u(i,j-1)^2-u(i-1,j-1)^2))/(8*h^2)
    end
    x=x+h
    end
end
x=0;
for i=1:6
    printf('
 %i	%.1f	|		',i-1,x)
    for j=1:7
        printf('%.3f	',u(i,j))
    end
    x=x+h
end
```

## Example 19.11: Finite_Difference_Method.sce

```scilab
//Example 19.11
//Finite Difference Method
//Page no. 670
clc;clear;close;
deff('y=f(x)','y=sin(%pi*x)')
deff('y=g(x)','y=0')
a=1;b=1;c=1;n=5;m=10;
h=a/n;k=b/m;r=c*k/h;
r1=r^2;r2=r1/2;s1=1-r1;s2=2*(1-r2)
printf('
 i ')
for i=1:n
    printf('	  %i',i)
end
printf('
-----------------------------------------------
fi')
for i=1:n
    f1(i)=f(h*(i-1))
    printf('	%.3f',f1(i))
end
printf('
gi')
for i=1:n
    g1(i)=g(h*(i-1))
    printf('	  %g',g1(i))
end
printf('


 i / j --> ')
for i=1:m
    printf('	  %i',i)
end
printf('
-------------------------------------------------------------------------------------------------')
for j=1:m
    for i=1:n
        if i==1 | i==n then
            u(i,j)=0;
        elseif j==1
            u(i,j)=f1(i)
        elseif j==2
            u(i,j)=s1*f1(i)+k*g1(i)+r2*(f1(i+1)+f1(i-1))
        else
            u(i,j)=s2*u(i,j-1)+r1*u(i-1,j-1)+u(i+1,j-1)-u(i,j-2)
        end
        
    end
end
for i=1:n
    printf('
 %i	',i)
    for j=1:m
        printf('	%.3f',u(i,j))
    end
end
```

## Example 19.3: Simple_Explicit_Method.sce

```scilab
//Example 19.3
//Simple Explicit Method
//Page no. 658
clc;clear;close;

c=-2;dt=0.07;dx=0.2;
r=abs(c)*dt/dx;
printf('
  x	i	|	j -->	')
for i=0:6
    printf('  %i	',i)
end
printf('
  |	|	|	t -->	')
for i=0:6
    printf('%.3f	',i*dt)
end
printf('
---------------------------------------------------------------------------------------')
for j=1:6
    printf('
 %.1f	%i	|		',(j-1)*dx,j-1)
    for i=1:7
        if i==1 then
            u(j,i)=0;
        elseif j==1 then
            u(j,i)=1
        else
            u(j,i)=(1-r)*u(j,i-1)+r*u(j-1,i-1)
        end
        printf('%.3f	',u(j,i))
    end
end
```

## Example 19.4: Simple_Implicit_Method.sce

```scilab
//Example 19.4
//Simple Implicit Method
//Page no. 659
clc;clear;close;

c=-2;dt=0.07;dx=0.2;
r=abs(c)*dt/dx;
printf('
  x	i	|	j -->	')
for i=0:6
    printf('  %i	',i)
end
printf('
  |	|	|	t -->	')
for i=0:6
    printf('%.3f	',i*dt)
end
printf('
---------------------------------------------------------------------------------------')
for j=1:6
    printf('
 %.1f	%i	|		',(j-1)*dx,j-1)
    for i=1:7
        if i==1 then
            u(j,i)=0;
        elseif j==1 then
            u(j,i)=1
        else
            u(j,i)=(1/(1+r))*u(j,i-1)+r*u(j-1,i)/(1+r)
        end
        printf('%.3f	',u(j,i))
    end
end
```

## Example 19.5: Lax_Wendroff_Method.sce

```scilab
//Example 19.5
//Lax Wendroff Method
//Page no. 660
clc;clear;close;

c=-2;dt=0.07;dx=0.2;
r=abs(c)*dt/dx;
printf('
  x	i	|	j -->	')
for i=0:6
    printf('  %i	',i)
end
printf('
  |	|	|	t -->	')
for i=0:6
    printf('%.3f	',i*dt)
end
i=1;
printf('
---------------------------------------------------------------------------------------')
for j=1:7
    for i=1:6
        if j==1 then
            u(i,j)=0;
            u(i+1,j)=0;
        elseif i==1 then
            u(i,j)=1
        else
            u(i,j)=r*(r-1)*u(i+1,j-1)/2+(1-r^2)*u(i,j-1)+r*(1+r)*u(i-1,j-1)/2
        end
    end
end
for i=1:6
    printf('
 %.1f	%i	|		',(i-1)*dx,i-1)
    for j=1:7
        printf('%.3f	',u(i,j))
    end
end
```

## Example 19.6: Wendroff_Method.sce

```scilab
//Example 19.6
//Wendroff Method
//Page no. 661
clc;clear;close;

c=2;k=0.07;h=0.2;
a=(h+k*c)/(h-k*c)
printf('
  x	i	|	j -->	')
for i=0:6
    printf('  %i	',i)
end
printf('
  |	|	|	t -->	')
for i=0:6
    printf('%.3f	',i*k)
end
printf('
---------------------------------------------------------------------------------------')
for i=1:6
    printf('
 %.1f	%i	|		',(i-1)*h,i-1)
    for j=1:7
        if j==1 then
            u(i,j)=0;
        elseif i==1 then
            u(i,j)=1
        else
            u(i,j)=u(i-1,j-1)+(u(i,j-1)-u(i-1,j))/a
        end
        printf('%.3f	',u(i,j))
    end
end
```

## Example 19.7: Leapfrog_Method.sce

```scilab
//Example 19.7
//Leapfrog Method
//Page no. 662
clc;clear;close;

c=2;k=0.07;h=0.2;
r=c*k/h
printf('
  x	i	|	j -->	')
for i=0:6
    printf('  %i	',i)
end
printf('
  |	|	|	t -->	')
for i=0:6
    printf('%.3f	',i*k)
end
printf('
---------------------------------------------------------------------------------------')

for j=1:7
    for i=1:6
        if j==1 | j==2 & i~=1 then
            u(i,j)=0;
            u(i+1,j)=0;
        elseif i==1 then
            u(i,j)=1
        else
            u(i,j)=u(i,j-2)-r*(u(i+1,j-1)-u(i-1,j-1))
        end
    end
end
for i=1:6
    printf('
 %.1f	%i	|		',(i-1)*h,i-1)
    for j=1:7
        printf('%.3f	',u(i,j))
    end
end
```

## Example 19.8: Variable_Coefficients.sce

```scilab
//Example 19.8
//Variable Coefficients
//Page no. 663
clc;clear;close;

//simple explicit method
printf('

By Simple Explicit Method:

')
dt=0.05;dx=0.2;
x=0;
printf('
 i	  x	  r	|	j -->	')
for i=0:6
    printf('  %i	',i)
end
printf('
  			|	t -->	')
for i=0:6
    printf('%.3f	',i*dt)
end
printf('
----------------------------------------------------------------------------------------------')
for j=1:6
    r=sqrt(1+2*x)*dt/dx;
    printf('
 %i	%.3f	%.3f	|		',(j-1),x,r)
    for i=1:7
        if i==1 then
            u(j,i)=0;
        elseif j==1 then
            u(j,i)=1
        else
            u(j,i)=(1-r)*u(j,i-1)+r*u(j-1,i-1)
        end
        printf('%.3f	',u(j,i))
        
    end
    x=x+dx
end


//simple implicit method
printf('


By Simple Implicit Method:
')
c=-2;dt=0.05;dx=0.2;x=0
printf('
 i	  x	  r	|	j -->	')
for i=0:6
    printf('  %i	',i)
end
printf('
  			|	t -->	')
for i=0:6
    printf('%.3f	',i*dt)
end
printf('
----------------------------------------------------------------------------------------------')
for j=1:6
    r=sqrt(1+2*x)*dt/dx;
    printf('
 %i	%.3f	%.3f	|		',(j-1),x,r)
    for i=1:7
        if i==1 then
            u(j,i)=0;
        elseif j==1 then
            u(j,i)=1
        else
            u(j,i)=(1/(1+r))*u(j,i-1)+r*u(j-1,i)/(1+r)
        end
        printf('%.3f	',u(j,i))
    end
    x=x+dx
end


//wendroff method
printf('


By Wendroff Method:
')
k=0.05;h=0.2;
x=0.1;
printf('
 i	  x	  c	  a	|	j -->	')
for i=0:6
    printf('  %i	',i)
end
printf('
  				|	t -->	')
for i=0:6
    printf('%.3f	',i*k)
end
printf('
-------------------------------------------------------------------------------------------------------')
for i=1:6
    c=sqrt(1+2*x);
    a=(h+k*c)/(h-k*c)
    printf('
 %i	%.3f	%.3f	%.3f	|		',(i-1),x-h/2,c,a)
    for j=1:7
        if j==1 then
            u(i,j)=0;
            u(i+1,j)=0;
        elseif i==1 then
            u(i,j)=1
        else
            u(i,j)=u(i-1,j-1)+(u(i,j-1)-u(i-1,j))/a
        end
        printf('%.3f	',u(i,j))
    end
    x=x+h
end
```

## Example 19.9: Inhomogeneous_1st_Order_Hyperboolic_Differential_Equation.sce

```scilab
//Example 19.9
//Inhomogeneous 1st Order Hyperboolic Differential Equation
//Page no. 665
clc;clear;close;

//simple explicit method
printf('

By Simple Explicit Method:
')
c=-2;dt=0.07;dx=0.2;
r=abs(c)*dt/dx;
printf('
  i	x	|	j -->	')
for i=0:6
    printf('  %i	',i)
end
printf('
  |	|	|	t -->	')
for i=0:6
    printf('%.3f	',i*dt)
end
printf('
---------------------------------------------------------------------------------------')
x=0;
for j=1:6
    printf('
 %i	%.1f	|		',j-1,x)
    for i=1:7
        if i==1 then
            u(j,i)=exp(-x);
        elseif j==1 then
            u(j,i)=1
        else
            u(j,i)=(1-r)*u(j,i-1)+r*u(j-1,i-1)+dt*2*x
        end
        printf('%.3f	',u(j,i))
    end
    x=x+dx
end


//simple implicit method
printf('


By Simple Implicit Method:
')
c=-2;dt=0.07;dx=0.2;
r=abs(c)*dt/dx;
printf('
  i	x	|	j -->	')
for i=0:6
    printf('  %i	',i)
end
printf('
  |	|	|	t -->	')
for i=0:6
    printf('%.3f	',i*dt)
end
printf('
---------------------------------------------------------------------------------------')
x=0;
for j=1:6
    printf('
 %i	%.1f	|		',j-1,x)
    for i=1:7
        if i==1 then
            u(j,i)=exp(-x);
        elseif j==1 then
            u(j,i)=1
        else
            u(j,i)=(1/(1+r))*u(j,i-1)+r*u(j-1,i)/(1+r)+dt*2*x
        end
        printf('%.3f	',u(j,i))
    end
    x=x+dx
end


//wendroff method
printf('


By Wendroff Method:
')
c=2;k=0.07;h=0.2;
a=(h+k*c)/(h-k*c)
printf('
  x	i	|	j -->	')
for i=0:6
    printf('  %i	',i)
end
printf('
  |	|	|	t -->	')
for i=0:6
    printf('%.3f	',i*k)
end
printf('
---------------------------------------------------------------------------------------')
x=0;
for i=1:6
    printf('
 %.1f	%i	|		',x,i-1)
    for j=1:7
        if j==1 then
            u(i,j)=exp(-x);
        elseif i==1 then
            u(i,j)=1
        else
            u(i,j)=u(i-1,j-1)+(u(i,j-1)-u(i-1,j))/a+(2*h*k)*(x+h/2)/(a*(h+c*k))
        end
        printf('%.3f	',u(i,j))
    end
    x=x+h
end
```
