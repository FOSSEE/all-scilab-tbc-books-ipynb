## Example 16.1: Chapter_16_Example_1.sce

```scilab
//Chapter 16,Example 1, page 556
//Determine the (a)ripple voltage (b)voltage drop (c)Average output volatge (d)ripple factor 
clear
clc
I1 = 5*10^-3 // A
C2 = 0.05*10^-6 // F
C1 = 0.01*10^-6 // F
Vs = 100 // kV
f = 50 // Hz 
// (a) Ripple voltage
printf('
 Part (a)')
delV = I1/(C2*f)
printf('
 Ripple Voltage = %f V', delV)
// (b) Voltage drop 
printf('
 Part (b)')
Vd = I1/f*((1/C1)+(1/(2*C2)))
printf('
 Voltage drop = %f V', Vd)
// (c) Average output voltage 
printf('
 Part (c)')
Vav = 2*Vs*sqrt(2)-Vd*10^-3
printf('
 Average output voltage = %f kV', Vav)
// (d) Ripple factor 
printf('
 Part (d)')
RF = Vd*10^-3/(2*Vs*sqrt(2)) 
printf('
 Ripple Factor in percentage = %f', RF*100)
```

## Example 16.2: Chapter_16_Example_2.sce

```scilab
//Chapter 16,Example 2, page 556
//Determine the (a)ripple voltage (b)voltage drop (c)Average output volatge (d)ripple factor 
clear
clc
I1 = 5*10^-3 // A
C3 = 0.10*10^-6 // F
C2 = 0.05*10^-6 // F
C1 = 0.01*10^-6 // F
Vs = 100 // kV
f = 50 // Hz 
// (a) Ripple voltage
printf('
 Part (a)')
delV = I1/f*((2/C1)+(1/C3))
printf('
 Ripple Voltage = %f kV', delV*10^-3)
// (b) Voltage drop 
printf('
 Part (b)')
Vd = I1/f*((1/C2)+(1/C1)+(1/(2*C3)))
printf('
 Voltage drop = %f kV', Vd*10^-3)
// (c) Average output voltage 
printf('
 Part (c)')
Vav = 3*Vs*sqrt(2)-Vd*10^-3
printf('
 Average output voltage = %f kV', Vav)
// (d) Ripple factor 
printf('
 Part (d)')
RF = Vd*10^-3/(3*Vs*sqrt(2)) 
printf('
 Ripple Factor in percentage = %f', RF*100)
// Answers may vary due to round off error
```

## Example 16.3: Chapter_16_Example_3.sce

```scilab
//Chapter 16,Example 3, page 557
//Determine the (a)ripple voltage (b)voltage drop (c)Average output volatge (d)ripple factor (e)optimum number of stages
clear
clc
I1 = 5*10^-3 // A
C = 0.15*10^-6 // F
Vs = 200 // kV
f = 50 // Hz 
n = 12
// (a) Ripple voltage
printf('
 Part (a)')
delV = I1*n*(n+1)/(f*C*2)
printf('
 Ripple Voltage = %f kV', delV*10^-3)
// (b) Voltage drop 
printf('
 Part (b)')
a = I1/(f*C)
Vd = a*((2/3*n^3)+(n^2/2)-(n/6)+(n*(n+1)/4))
printf('
 Voltage drop = %f kV', Vd*10^-3)
// (c) Average output voltage 
printf('
 Part (c)')
Vav = 2*n*Vs*sqrt(2)-Vd*10^-3
printf('
 Average output voltage = %f kV', Vav)
// (d) Ripple factor 
printf('
 Part (d)')
RF = Vd*10^-3/(2*n*Vs*sqrt(2)) 
printf('
 Ripple Factor in percentage = %f', RF*100)
// (e) Optimum number of stages
printf('
 Part (e)')
nopt = sqrt(Vs*sqrt(2)*10^3*f*C/I1) 
printf('
 Optimum number of stages = %d stages', nopt)
// Answers may vary due to round off error
```

## Example 16.4: Chapter_16_Example_4.sce

```scilab
//Chapter 16,Example 4, page 558
//Determine the input voltage and power
clear
clc
Vc = 500*10^3 // V
A = 4 // A
Xl = 8/100 // in percentage 
kV = 250
Xc = Vc/A // Reactance of the cable
XL = Xl*(kV**2/100)*10**3 // Leakage reactance of the transformer
Radd = Xc-XL // Additional series reactance
Ind = Radd/(2*%pi*XL) // Inductance of required series inductor
R = 3.5/100*(kV**2/100)*10**3  // Total circuit resistance
Imax = 100/250 // maximum current that can be supplied by the transformer
Vex = Imax*R // Exciting voltage of transformer secondary
Vin = Vex*220/kV // Input voltage of transformer primary
P = Vin*100/220 // Input power of the transformer
printf('
 Reactance of the cable = %f k ohm', Xc*10^-3)
printf('
 Leakage reactance of the transformer = %f k ohm', XL*10^-3)
printf('
 Additional series reactance = %f k ohm', Radd*10^-3)
printf('
 Inductance of required series inductor = %f H', Ind*10^3)
printf('
 Total circuit resistance = %f k ohm', R*10^-3)
printf('
 maximum current that can be supplied by the transformer = %f A', Imax)
printf('
 Exciting voltage of transformer secondary = %f kV', Vex*10^-3)
printf('
 Input voltage of transformer primary = %f V', Vin*10^-3)
printf('
 Input power of the transformer = %f kW', P*10^-3)
// Answers may vary due to round off error
```
