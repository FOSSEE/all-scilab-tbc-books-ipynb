## Example 3.10: Designing_a_Unity_Gain_Low_Pass_Filter.sce

```scilab
//Example 3.10

clear;

clc;

C=1*10^(-9);

Q=2;

n=(4*Q^(2))+4;

C1=n*C;

C2=C;

f0=10*10^(3);

k=(n/(2*(Q^(2))))-1;

m=k+((k^2-1)^(0.5));

k1=(m*n)^(0.5);

R=1/(k1*2*%pi*f0*C);

R2=R;

R1=m*R;

printf('Designed Unity Gain Low Pass Filter :');

printf('
R1=%.2f kohms',R1*10^(-3));

printf('
R2=%.2f kohms',R2*10^(-3));

printf('
C1=%.f nF',C1*10^9);

printf('
C2=%.f nF',C2*10^9);
```

## Example 3.11_a: Designing_Butterworth_Low_Pass_Filter.sce

```scilab
//Example 3.11(a)

clear;

clc;

m=1;//Q is maximised at m=1

n=2;//Order of filter

f0=10*10^(3);

Qnum=(m*n)^(1/2);

Qden=m+1;

Q=Qnum/Qden;

C=1*10^(-9);//Assuming C=1 nF

C2=C;

C1=n*C;

R=1/(Qnum*C*2*%pi*f0);

R2=R;

R1=m*R;

printf('Designed Second Order Low Pass Butterworth Filter :')

printf('
R1=%.2f kohms',R1*10^(-3));

printf('
R2=%.2f kohms',R2*10^(-3));

printf('
C1=%.f nF',C1*10^9);

printf('
C2=%.f nF',C2*10^9);
```

## Example 3.13_a: Designing_KRC_Bandpass_Filter.sce

```scilab
//Example 3.13(a)

clear;

clc;

C=10*10^(-9);//Assumed

C1=C;

C2=C;

f0=1*10^3;

BW=100;

R=(2^(1/2))/(2*%pi*f0*C);

R1=R;

R2=R;

R3=R;

Q=f0/BW;

K=4-((2^(1/2))/Q);

RA=10*10^3;

RB=(K-1)*RA;

RG=K/(4-K);

printf('Designed KRC Second Order Band Pass filter');

printf('
R1=R2=R3=%.1f kohms',R*10^(-3));

printf('
RA=%.2f kohms',RA*10^(-3));

printf('
RB=%.2f kohms',RB*10^(-3));

printf('
C1=C2=%.2f nF',C*10^9);

printf('

Resonance Gain=%.2f V/V',RG);
```
