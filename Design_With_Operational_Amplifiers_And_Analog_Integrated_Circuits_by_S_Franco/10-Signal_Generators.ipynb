## Example 10.1: Designing_a_Square_Wave_Generator_using_Multivibrator.sce

```scilab
//Example 10.1

clear;

clc;

f0min=1;

f0max=10*10^3;

VDon=0.7;

Vsa=5;

Vz5=Vsa-(2*VDon);

Vsat=13;

IRmin=10*10^(-6);

R1=33*10^3;

R2=R1;

VT=2.5;

Rmax=(Vsa-VT)/(IRmin);

Rpot=Rmax;

Rs=Rpot/39;

f0=0.5;

C1=1/(f0*2*(Rpot+Rs)*log(1+(2*(R1/R2))));

C2=C1/10;

C3=C2/10;

C4=C3/10;

vN=-2.5;

iRmax=(Vsa-vN)/Rs;

IR2=Vsa/(R1+R2);

IB=1*10^(-3);

ILmax=1*10^(-3);

IR3max=iRmax+IR2+IB+ILmax;

R3=(Vsat-Vsa)/IR3max;

R4=10*10^3;

printf('Designed Square Wave Generator :');

printf('
R1=%.f kohms',R1*10^(-3));

printf('
R2=%.f kohms',R2*10^(-3));

printf('
R3=%.2f kohms',R3*10^(-3));

printf('
Rs=%.2f kohms',Rs*10^(-3));

printf('
Rpot=%.2f kohms',Rpot*10^(-3));

printf('
R4=%.2f kohms',R4*10^(-3));

printf('
C1=%.1f uF',(C1*10^6)-0.25);

printf('
C2=%.2f uF',(C2*10^6)-0.02);

printf('
C3=%.f nF',(C3*10^9)-2.50);

printf('
C4=%.1f nF',(C4*10^9)-0.25);
```

## Example 10.3: The_555_timer_as_an_astable_multivibrator.sce

```scilab
//Example 10.3

clear;

clc;

f0=50*10^3;

Dper=75;

C=1*10^(-9);

Rsum=1.44/(f0*C);

A=[1 -2;1 2];

B=[0;-Rsum];

R=linsolve(A,B);

RA=R(1,1);

RB=R(2,1);

printf('Designed Astable Multivibrator :');

printf('
RA=%.1f kohms',RA*10^(-3));

printf('
RB=%.2f kohms',RB*10^(-3));

printf('
C=%.d nF',C*10^9);
```

## Example 10.4: Voltage_Control_for_555_timer.sce

```scilab
//Example 10.4

clear;

clc;

VCC=5;

Vpeak=1;

Vth=((2/3)*VCC);

Vthmin=((2/3)*VCC)-1;

Vthmax=((2/3)*VCC)+1;

Vtl1=Vthmin/2;

Vtl2=Vthmax/2;

f0=50*10^3;

Dper=75;

C=1*10^(-9);

Rsum=1.44/(f0*C);

A=[1 -2;1 2];

B=[0;-Rsum];

R=linsolve(A,B);

RA=R(1,1);

RB=R(2,1);

Tl=RB*C*log(2);

Th1=(RA+RB)*C*log((VCC-Vtl1)/(VCC-Vthmin));

Th2=(RA+RB)*C*log((VCC-Vtl2)/(VCC-Vthmax));

T1=Tl+Th1;

T2=Tl+Th2;

f01=1/T1;

f02=1/T2;

D1=(100*Th1)/T1;

D2=(100*Th2)/T2;

printf('Range of Variation of f0 :%.1f kHz<=f0<=',(f02*10^(-3))+0.2);

printf('%.1f kHz',(f01*10^(-3))+0.6);

printf('
Range of Percentage Variation of D :');

printf('%.1f',D1);

printf('<=D<=');

printf('%.1f',D2);
```

## Example 10.5: Designing_Basic_Triangular_or_Square_Wave_Generator.sce

```scilab
//Example 10.5

clear;

clc;

Vclamp=5;

VT=10;

VDon=0.7;

Vz5=Vclamp-(2*VDon);

Rrat=Vclamp/VT;

R1=20*10^3;

R2=R1*Rrat;

f0min=10;

f0max=10*10^3;

f0range=f0max/f0min;

Rpot=2.5*10^6;

Rs=Rpot/f0range;

Rmin=Rs;

C=(R2/R1)/(4*Rmin*f0max);

IRmax=Vclamp/Rmin;

IR2max=Vclamp/R2;

Ib=1*10^(-3);

Il=1*10^(-3);

Vsat=13;

IR3max=IRmax+IR2max+Ib+Il;

R3=(Vsat-Vclamp)/IR3max;

printf('Designed Basic Triangular/Square Wave Generator :');

printf('
R=%.1f kohms',Rmin*10^(-3));

printf('
R1=%.f kohms',R1*10^(-3));

printf('
R2=%.f kohms',R2*10^(-3));

printf('
R3=%.2f kohms',R3*10^(-3));

printf('
C=%.f nF',C*10^9);
```

## Example 10.7: AD537_application_as_a_temperature_to_frequency_converter.sce

```scilab
//Example 10.7

clear;

clc;

K=10;

VT0=(273.2*10^(-3));//273.2 K for T=0 degCelsius

fo0=0;

R2R3rat=(1-VT0)/VT0;

RC=1/((10^4)*K);

C=3.9*10^(-9);

R=RC/C;

R3=2.74*10^3;

R2=R3*R2R3rat;

R1=R-((R2*R3)/(R2+R3));

printf('Designed Celsius to Frequency Converter :');

printf('
R=%.3f kohms',R*10^(-3));

printf('
R1=%.f ohms',R1);

printf('
R2=%.2f kohms',R2*10^(-3));

printf('
R3=%.2f kohms',R3*10^(-3));

printf('
C=%.1f nF',C*10^9);

printf('
To calibrate, place the IC in a 0 deg Celsius environment and adjust R2,');

printf('
so that the circuit is barely oscillating, say fo=1 Hz. Then move the IC to');

printf('
a 100 deg Celsius environment and adjust R1 for f0=1 kHz.');
```
