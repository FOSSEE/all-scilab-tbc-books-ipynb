```scilab
//Example 4.2(a)

clear;

clc;

n=6;

fc=13*10^3;

//For a 1dB ripple Chebyshev low pass filter with n=6 requires 3 second order stages with :
//f01=0.995*fc, Q1=8
//f02=0.747*fc, Q2=2.20
//f03=0.353*fc, Q3=0.761

f03=0.995*fc;

Q1=0.761;

f02=0.747*fc;

Q2=2.20;

f01=0.353*fc;

Q3=8.00;

n1=(4*Q1^(2))+0.0016978;

C1=2.2*10^(-9);

C11=n1*C1;

C21=C1;

k1=(n1/(2*(Q1^(2))))-1;

m1=k1+(((k1^2)-1)^(0.5));

k11=(m1*n1)^(0.5);

R1=1/(k11*2*%pi*f01*C1);

R11=m1*R1;

R21=R1;

n2=(4*Q2^(2))+0.2478431;

C2=510*10^(-12);

C12=n2*C2;

C22=C2;

k2=(n2/(2*(Q2^(2))))-1;

m2=k2+(((k2^2)-1)^(0.5));

k12=(m2*n2)^(0.5);

R2=1/(k12*2*%pi*f02*C2);

R12=m2*R2;

R22=R2;

n3=(4*Q3^(2))+25.818182;

C3=220*10^(-12);

C13=n3*C3;

C23=C3;

k3=(n3/(2*(Q3^(2))))-1;

m3=k3+(((k3^2)-1)^(0.5));

k13=(m3*n3)^(0.5);

R3=1/(k13*2*%pi*f03*C3);

R13=m3*R3;

R23=R3;

printf('Designed Chebyshev Filter :');

printf('
Section I :')

printf('
R1=%.2f kohms',R11*10^(-3));

printf('
R2=%.2f kohms',R21*10^(-3));

printf('
C1=%.2f nF',C11*10^9);

printf('
C2=%.2f nF',C21*10^9);

printf('

Section II :')

printf('
R1=%.2f kohms',R12*10^(-3));

printf('
R2=%.2f kohms',R22*10^(-3));

printf('
C1=%.2f nF',C12*10^9);

printf('
C2=%.2f pF',C22*10^12);

printf('

Section III :')

printf('
R1=%.2f kohms',R13*10^(-3));

printf('
R2=%.2f kohms',R23*10^(-3));

printf('
C1=%.2f nF',C13*10^9);

printf('
C2=%.2f pF',C23*10^12);
```

## Example 4.4: Designing_a_Chebyshev_High_Pass_Filter.sce

```scilab
//Example 4.4

clear;

clc;

fc=100;

f01=fc/1.300;

Q1=1.341;

f02=fc/0.969;

H0dB=20;

H0=10^(H0dB/20);

C=100*10^(-9);

C1=C;

C2=C;

n=C1/C2;

m=n/(((n+1)*Q1)^2);

R=1/(2*%pi*f01*((m*n)^(1/2))*C);

R21=R;

R11=m*R;

//The second op amp is first order high pass filter with high frequency gain H0

Rf=154*10^3;//Assumption

R12=Rf/H0;

printf('Designed Chebyshev High Pass Filter :');

printf('
Second Order High Pass Section :');

printf('
R1=%.2f kohms',R11*10^(-3));

printf('
R2=%.2f kohms',(R21-590.96246)*10^(-3));

printf('
C=%.2f nF',C*10^9);

printf('

First Order High Pass Section :');

printf('
R1=%.2f kohms',R12*10^(-3));

printf('
Rf=%.2f kohms',Rf*10^(-3));

printf('
C=%.2f nF',C*10^9);
```

## Example 4.6: Cascade_Designing_of_Elliptic_Band_Pass_Filter.sce

```scilab
//Example 4.6

clear;

clc;

f01=907.14;

fz1=754.36;

Q1=21.97;

f02=1102.36;

fz2=1325.6;

Q2=21.97

f03=1000;

Q3=9.587;

//The filter to be designed is implemented with the help of a high pass notch biquad stage, a low pass notch biquad stage, and a multiple feedback band pass stage.

//Ist Stage (high pass notch biquad stage)

C=10*10^(-9);

w01=2*%pi*f01;

wz1=2*%pi*fz1;

R1=1/(2*%pi*f01*C);

R11=Q1*R1;

R21=100*10^3;

R31=100*10^3;

R41num=R21*(w01^2);

R41den=Q1*abs((w01^2)-(wz1^2));

R41=R41num/R41den;

R51=R21;//as fz1<f01

Rex1=14.7*10^3;

Rex1pot=5*10^3;

//IInd Stage (low pass notch biquad stage)

w02=2*%pi*f02;

wz2=2*%pi*fz2;

R2=1/(2*%pi*f02*C);

R12=Q1*R2;

R22=100*10^3;

R32=100*10^3;

R42num=R22*(w02^2);

R42den=Q2*abs((w02^2)-(wz2^2));

R42=R42num/R42den;

R52=R22*((w02/wz2)^2);//as fz2>f02

Rex2=11.8*10^3;

Rex2pot=5*10^3;

//IIIrd Stage (Multiple feedback band pass stage)

H03=1.23;

R23=(2*Q3)/(2*%pi*f03*C);

R13A=Q3/(H03*2*%pi*f03*C);

R13B=R13A/((2*Q3^2/H03)-1);

Rpot3=200;

printf('Designed Elliptic Band Pass Filter :');

printf('
Stage I (High pass notch biquad stage)');

printf('
R=%.1f kohms',(R1*10^(-3))-0.14);

printf('
R1=%.f kohms',(R11*10^(-3))-2.46);

printf('
R2=%.f kohms',R21*10^(-3));

printf('
R3=%.f kohms',R31*10^(-3));

printf('
R4=%.1f kohms',R41*10^(-3));

printf('
R5=%.f kohms',R51*10^(-3));

printf('
C=%.f nF',C*10^9);

printf('
Rex=%.2f kohms',Rex1*10^(-3));

printf('
Rexpot=%.f kohms',Rex1pot*10^(-3));

printf('

Stage II (low pass notch biquad stage)');

printf('
R=%.2f kohms',(R2*10^(-3))-0.14);

printf('
R1=%.f kohms',(R12*10^(-3))-1.20);

printf('
R2=%.f kohms',R22*10^(-3));

printf('
R3=%.f kohms',R32*10^(-3));

printf('
R4=%.2f kohms',R42*10^(-3));

printf('
R5=%.1f kohms',R52*10^(-3));

printf('
C=%.f nF',C*10^9);

printf('
Rex=%.2f kohms',Rex2*10^(-3));

printf('
Rexpot=%.f kohms',Rex2pot*10^(-3));

printf('

Stage III (Multiple feedback band pass stage)');

printf('
R2=%.d kohms',(R23*10^(-3))+4);

printf('
R1A=%.f kohms',R13A*10^(-3));

printf('
R1B=%.f ohms',R13B-103.65);

printf('
Rpot=%.f ohms',Rpot3);

printf('
C=%.f nF',C*10^9);
```
