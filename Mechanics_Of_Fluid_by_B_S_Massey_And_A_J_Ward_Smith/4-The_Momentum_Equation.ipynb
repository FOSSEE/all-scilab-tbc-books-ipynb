# Chapter 4: The Momentum Equation

## Example 4.1: 1.sce

```scilab
clc

rho=1000; // kg/m^3
u1=36; // m/s
u2=30; // m/s
d=0.05; // m
theta=60; // degrees

A=%pi/4*d^2;

Q=A*u1;

F_x=rho*Q*(u2*cosd(theta) - u1);
F_y=rho*Q*u2*sind(theta);

F=sqrt(F_x^2+F_y^2);
phi=atand(F_y/F_x);

disp('The Hydrodynamic force on the vane =')
disp(F)
disp('N')

printf('This resultant force acts at angle of %f to the x-direction', phi)
 
```

## Example 4.2: 2.sce

```scilab
clc

Q1=0.45; // m^3/s
Q2=0.425; // m^3/s
d1=0.6; // m
d2=0.3; // m
p1=1.4*10^5; // Pa
rho=1000; // kg/m^3
theta=45; // degrees

A1=%pi/4*d1^2;
A2=%pi/4*d2^2;

u1=Q1/A1;
u2=Q2/A2;

p2=p1+rho/2*(u1^2-u2^2);

F_x=rho*Q2*(u2*cosd(theta)-u1)-p1*A1+p2*A2*cosd(theta)
F_y=rho*Q2*(u2*sind(theta)-0)+p2*A2*sind(theta);

F=sqrt(F_x^2+F_y^2);
phi=atand(F_y/F_x);

disp('The net horizontal force exerted by the water onthe bend =')
disp(F)
disp('N')

printf('This resultant force acts at angle of %f to the x-direction', phi)
```

## Example 4.3: 3.sce

```scilab
clc

rho=1.2; // kg/m^3
d=12; // m
u1=20; // m/s
u4=8; // m/s

A=%pi/4*d^2
F=rho*A*(u1+u4)/2*(u1-u4);

disp('(b) The thrust on the turbine = ')
disp(F)
disp('N')

P=rho*A*(u1+u4)/2*(u1^2/2-u4^2/2);
disp('Power generated by the turbine =')
disp(P)
disp('W')
```
